```python
def get_title(vname_in):
    '''
    Return the xlabel string consisting of 
    varname and unit
    '''
    var_unit = pu.formatting.get_unit_mplstr(bp12.model_utils.get_varinfo(vname_in, 'unit'))
    var_longname = bp12.model_utils.get_varinfo(vname_in,'long_name')
    return f"{var_longname} {var_unit}"
```

### Functions 

```python
def plt_profiles(annstr, var_obs, var_mdl, mld_obs, mld_mdl, xlabelin, maxdepth=None, prec=None, minmax=None):
    if prec is None: prec=2
    if not maxdepth: maxdepth=400
    if 'month' in var_obs.dims: var_obs = var_obs.rename({'month':'time'})
    if 'month' in var_mdl.dims: var_mdl = var_mdl.rename({'month':'time'})
    
    allvals = np.concatenate((var_obs.sel(depth=slice(0,maxdepth)).values.flatten(),
                              var_mdl.sel(deptht=slice(0,maxdepth)).values.flatten())
                            )
    if minmax:
        xminp, xmaxp = minmax
    else:
        xminp, xmaxp = pu.formatting.get_xylim(np.nanmin(allvals), np.nanmax(allvals), prec)
    
    
    if 'time' in var_obs.dims:
        mean_obs, m1_obs, m9_obs = var_obs.mean('time'), var_obs.isel(time=0), var_obs.isel(time=8)
        ax.plot(m1_obs, -m1_obs.depth, 'b:', alpha=0.6)
        ax.plot(m9_obs, -m9_obs.depth, 'b--', alpha=0.6)
        ax.fill_betweenx(-m1_obs.depth.values, m1_obs.values, m9_obs.values,
                         color='b', alpha=0.2)
    else: 
        mean_obs = var_obs
    ax.plot(mean_obs, -mean_obs.depth, 'b', label='OBS')
    
    mean_mdl, m1_mdl, m9_mdl = var_mdl.mean('time'), var_mdl.isel(time=0), var_mdl.isel(time=8)
    ax.plot(mean_mdl, -mean_mdl.deptht, 'r', label='BIOPERIANT12')
    ax.plot(m1_mdl, -m1_mdl.deptht, 'r:', alpha=0.6)
    ax.plot(m9_mdl, -m9_mdl.deptht, 'r--', alpha=0.6)
    ax.fill_betweenx(-m1_mdl.deptht.values, m1_mdl.values, m9_mdl.values,
                             color='r', alpha=0.2) 
    ax.set_xlim([xminp, xmaxp])
    xmin, xmax, _, _ = ax.axis()
    ax.plot([xmin, xmax], [-1*mld_obs.mean('time'), -1*mld_obs.mean('time')], 'b') 
    ax.plot([xmin, xmax], [-1*mld_obs[0], -1*mld_obs[0]], 'b:',   lw=2, alpha=0.4) 
    ax.plot([xmin, xmax], [-1*mld_obs[8], -1*mld_obs[8]], 'b--', lw=2, alpha=0.4)
    ax.plot([xmin, xmax], [-1*mld_mdl.mean('time'), -1*mld_mdl.mean('time')], 'r') 
    ax.plot([xmin, xmax], [-1*mld_mdl[0], -1*mld_mdl[0]], 'r:',   lw=2, alpha=0.4) 
    ax.plot([xmin, xmax], [-1*mld_mdl[8], -1*mld_mdl[8]], 'r--', lw=2, alpha=0.4)
    
  
    ax.set_xlabel(xlabelin, fontsize=12)
    ax.set_ylim([-1*maxdepth,0])    
    ax.set_ylabel('Depth [m]', fontsize=12)
    ax.grid(lw=1, c='0.75', alpha=0.4)
    plt.title(annstr, loc='left', x=-0.3, y=1.03, fontsize=12)
    
def vadd_tlegend():
    '''add legend for times'''
    lmean = mpl.lines.Line2D([], [], color='k', label='Mean')
    lm1 = mpl.lines.Line2D([], [], color='k', alpha=0.8, linestyle=':',  label='Jan')
    lm2 = mpl.lines.Line2D([], [], color='k', alpha=0.8, linestyle='--', label='Sep')
    leg2 = ax.legend(handles=[lmean, lm1, lm2], fontsize=12, frameon=False, bbox_to_anchor=(0, 1.1),
                   loc="lower left", ncol=3)
    ax.add_artist(leg2)
    
def vadd_omlegend():
    '''add legend forobs/model'''
    line_mdl = mpl.lines.Line2D([], [],  color='r', marker='o', linestyle='None', markersize=10, label='BIOPERIANT12')
    line_obs  = mpl.lines.Line2D([], [], color='b', marker='o', linestyle='None', markersize=10, label='OBS')
    leg1 = ax.legend(handles = [line_obs,line_mdl], ncol=2, fontsize=12, bbox_to_anchor=(0, 1.1),
                     frameon=False, loc="lower left")
    ax.add_artist(leg1)  
    
```
