## PLOT

```python
cmapDiff = cmo.tools.crop_by_percent(mpl.cm.RdBu_r, 1, which='both', N=None) 
normDiff = mpl.colors.BoundaryNorm(np.arange(-50,60,5), cmapDiff.N, extend='both')
```

```python
cmapDIC = cmo.tools.crop_by_percent(cm.bamO, 10, which='both', N=None) 
normDIC = mpl.colors.BoundaryNorm(np.arange(2000,2230,10), cmapDIC.N, extend='both')

cmapAlk = cmo.tools.crop_by_percent(cm.bamO, 10, which='both', N=None) 
normAlk = mpl.colors.BoundaryNorm(np.arange(2240,2420,10), cmapAlk.N, extend='both')

fig = plt.figure(figsize=(12, 6))

### ---------DIC ---MODEL--OBS---BIAS------------------------------------------- ###
ax = fig.add_subplot(2, 3, 1, projection = ccrs.Stereographic(central_latitude=-90))
plt.title(f'(a) BIOPERIANT12 DIC', loc='left',fontsize=12)
mesh = ax.pcolormesh(dic_0_mdl.lon, dic_0_mdl.lat, dic_0_mdl, 
                     cmap=cmapDIC, norm=normDIC,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(DIC_cbar, rotation=-90, labelpad=15)
pu.map_plots.map_decorator(ax)
pu.map_plots.add_biomes(ax, "mdl")

ax2 = fig.add_subplot(2, 3, 2, projection = ccrs.Stereographic(central_latitude=-90))
plt.title(f'(b) OBS DIC', loc='left',fontsize=12)
mesh = ax2.pcolormesh(dic_0_mdl.lon, dic_0_mdl.lat, dic_0_mdl, 
                     cmap=cmapDIC, norm=normDIC,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(DIC_cbar, rotation=-90, labelpad=15)
pu.map_plots.map_decorator(ax2)
pu.map_plots.add_biomes(ax2, "obs")

### ---------TA---MODEL--OBS---BIAS------------------------------------------- ###
ax4 = fig.add_subplot(2, 3, 4, projection = ccrs.Stereographic(central_latitude=-90))
plt.title(f'(c) BIOPERIANT12 TA', loc='left',fontsize=12)
mesh = ax4.pcolormesh(alk_0_obs.lon, alk_0_obs.lat, alk_0_obs, 
                     cmap=cmapAlk, norm=normAlk,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(ALK_cbar, rotation=-90, labelpad=15)
pu.map_plots.map_decorator(ax4)
pu.map_plots.add_biomes(ax4, "mdl")

ax5 = fig.add_subplot(2, 3, 5, projection = ccrs.Stereographic(central_latitude=-90))
plt.title(f'(d) OBS TA', loc='left',fontsize=12)
mesh = ax5.pcolormesh(alk_0_obs.lon, alk_0_obs.lat, alk_0_obs, 
                     cmap=cmapAlk, norm=normAlk,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(ALK_cbar, rotation=-90, labelpad=15)
pu.map_plots.map_decorator(ax5)
pu.map_plots.add_biomes(ax5, "obs")

plt.savefig(f'../figures/DIC1.png', bbox_inches='tight', dpi=400)
```

## Plot timeseries

```python
color15, color16, color17 = [pu.formatting_plots.get_biome_colors(bnum) for bnum in [15,16,17]]
```

```python
dic_ylabel = 'DIC [\u03BCmol.kg$^{-1}$]\n solid line'
alk_ylabel ='TA [\u03BCmol.kg$^{-1}$]\n dashed line'
```

```python
from mpl_toolkits.axes_grid.anchored_artists import AnchoredText
```

```python
def plot_ts_2ax(dic_in, dic_minmax, alk_in, alk_minmax, tsclr):
    
    dic_tick, alk_tick = dic_minmax[2], alk_minmax[2]
    ax.plot(dic_in.time, dic_in.values, label='BIOPERIANT12', c=tsclr, lw=2)
    ax.set_xlim([pd.Timestamp(1999,12,31), pd.Timestamp(2010,1,1)])
    ax.set_ylim([dic_minmax[0], dic_minmax[1]])
    ax.set_yticks(np.arange(dic_minmax[0], dic_minmax[1]+dic_tick, dic_tick))
    ax.set_ylabel(DIC_cbar, labelpad=5, fontsize=12)

    
    #add_source(ax)
    #add_var(ax)
    ax.grid()
    ax.tick_params(labelsize=12)
    ax.set_rasterized(True)
    ###--- ax2
    ax2 = ax.twinx()
    ax2.plot(alk_in.time, alk_in.values,  c=tsclr, linestyle='--', lw=2)
    ax2.set_ylim([alk_minmax[0], alk_minmax[1]])

    ax2.set_yticks(np.arange(alk_minmax[0], alk_minmax[1]+alk_tick, alk_tick))
    _, _, ymin, ymax = ax2.axis()
    pu.timeseries_plots.add_seas(ax2, (ymin, ymax), False)
    ax2.set_ylabel(ALK_cbar, labelpad=25, fontsize=12, rotation=-90)
    ax2.set_rasterized(True)
    
def add_var(axin, posin, colrin):
    line_dic = mpl.lines.Line2D([], [], color=colrin, lw=2, label='DIC')
    line_alk = mpl.lines.Line2D([], [], color=colrin, lw=2, linestyle='--', label='TA')
    leg1 = axin.legend(handles = [line_dic, line_alk], ncol=2, fontsize=12,
                  bbox_to_anchor=posin, loc="lower left",frameon=False)
    axin.add_artist(leg1)
    
def add_obs(dic_obs,alk_obs):
    obstr=f"OBS: DIC={dic_obs:3.2f}, TA={alk_obs:3.2f}"
    ax.annotate(obstr, xy=(0.7, 1.1), xycoords="axes fraction", bbox=dict(boxstyle="round", fc="w"))
```

### Surface

```python
def add_obs(dic_obs,alk_obs):
    obstr=f"OBS: DIC={dic_obs:3.2f}, TA={alk_obs:3.2f}"
    ax.annotate(obstr, xy=(0.55, 1.1), xycoords="axes fraction", bbox=dict(boxstyle="round", fc="w"))

fig, axs = plt.subplots(4, 6, constrained_layout=True, figsize=(8, 6))
gs = axs[0, 0].get_subplotspec().get_gridspec()
[a.remove() for i in range(6) for a in axs[:, i]]
        
ax = fig.add_subplot(gs[:1, 1:5])
plot_ts_2ax(dic_bts_mdl.DIC17_mean, (2050, 2250, 40), alk_bts_mdl.Alkalini17_mean, (2260, 2345,20), color17)
plt.title(f'(e) SO-ICE BIOPERIANT12',loc='left',fontsize=12)
add_obs(float(dic_bts_obs.DIC17_mean.values), float(alk_bts_obs.TAlk17_mean.values))
add_var(ax, (0., 0.68), color17)

ax = fig.add_subplot(gs[1:2, 1:5])
plot_ts_2ax(dic_bts_mdl.DIC16_mean, (2140, 2160, 5), alk_bts_mdl.Alkalini16_mean, (2290, 2305, 4), color16)
plt.title(f'(f) SO-SPSS BIOPERIANT12',loc='left',fontsize=12)
add_obs(float(dic_bts_obs.DIC16_mean.values), float(alk_bts_obs.TAlk16_mean.values))

ax = fig.add_subplot(gs[2:3, 1:5])
plot_ts_2ax(dic_bts_mdl.DIC15_mean, (2070, 2110, 10), alk_bts_mdl.Alkalini15_mean, (2310, 2322, 3), color15)
plt.title(f'(g) SO-STSS BIOPERIANT12',loc='left',fontsize=12)
add_obs(float(dic_bts_obs.DIC15_mean.values), float(alk_bts_obs.TAlk15_mean.values))
ax.set_xlabel('Year', labelpad=5, fontsize=12)
plt.savefig(f'../figures/DIC2.png', bbox_inches='tight', dpi=400)
```

```python
def plot_ts(axin, mdlin, stdin, yminmax, tsclr, ylbl):
    
    ymin_in, ymax_in, dy_in = yminmax[0], yminmax[1], yminmax[2]
    axin.plot(mdlin.time, mdlin.values, c=tsclr, lw=2)
    axin.fill_between(mdlin.time, mdlin.values-stdin.values, mdlin.values+stdin.values, color=tsclr, alpha=0.4)
    axin.set_xlim([pd.Timestamp(1999,12,31), pd.Timestamp(2010,1,1)])
    axin.set_ylim([ymin_in, ymax_in])
    yticks    = np.arange(ymin_in, ymax_in+dy_in, dy_in)
    yticklbls = [yt if i%2==0 else "" for i,yt in enumerate(yticks)]
    axin.set_yticks(yticks)
    axin.set_yticklabels(yticklbls)
    axin.set_ylabel(ylbl, labelpad=5, fontsize=12)
    _, _, ymin, ymax = axin.axis()
    pu.timeseries_plots.add_seas(axin, (ymin, ymax), False)
    
    axin.grid()
    axin.tick_params(labelsize=12)
    axin.set_rasterized(True)
    axin.set_xlabel('Year', labelpad=10, fontsize=12)
#
```

```python
fig, axs = plt.subplots(4, 6, constrained_layout=True, figsize=(12, 6))
gs = axs[0, 0].get_subplotspec().get_gridspec()
[a.remove() for i in range(6) for a in axs[:, i]]
        
ax = fig.add_subplot(gs[:1, 1:5])
ax.plot([dic_bts_mdl.time[0].values,dic_bts_mdl.time[-1].values],
         [dic_bts_obs.DIC17_mean, dic_bts_obs.DIC17_mean], color17, linestyle='--', linewidth=3)
plot_ts(ax, dic_bts_mdl.DIC17_mean, dic_bts_mdl.DIC17_std, (2100, 2240, 20), color17, DIC_cbar)
plt.title(f'(a) DIC SO-ICE',loc='left',fontsize=12)


ax = fig.add_subplot(gs[1:2, 1:5])
ax.plot([dic_bts_mdl.time[0].values,dic_bts_mdl.time[-1].values],
         [dic_bts_obs.DIC16_mean, dic_bts_obs.DIC16_mean], color16, linestyle='--', linewidth=3)
plot_ts(ax, dic_bts_mdl.DIC16_mean, dic_bts_mdl.DIC16_std, (2100, 2200, 20), color16, DIC_cbar)
plt.title(f'(c) DIC SO-SPSS',loc='left',fontsize=12)

ax = fig.add_subplot(gs[2:3, 1:5])
ax.plot([dic_bts_mdl.time[0].values,dic_bts_mdl.time[-1].values],
         [dic_bts_obs.DIC15_mean, dic_bts_obs.DIC15_mean], color15, linestyle='--', linewidth=3)
plot_ts(ax, dic_bts_mdl.DIC15_mean, dic_bts_mdl.DIC15_std, (2020, 2140, 20), color15, DIC_cbar)
plt.title(f'(e) DIC SO-STSS',loc='left',fontsize=12)
ax.set_xlabel('Year', labelpad=5, fontsize=12)
#plt.savefig(f'{evaldir}/FIGURES/ALKDIC1.png', bbox_inches='tight', dpi=400)
```
