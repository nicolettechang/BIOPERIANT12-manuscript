# Plot maps of bloom initiation, termination and duration from climatology

Code: T.J. Ryan-Keogh

```python
import sys
import os
src_path = os.path.abspath(os.path.join(os.getcwd(), '../src'))
sys.path.append(src_path)
import bp12_tools as bp12
import bp12_tools.plot_utils as pu

import numpy as np
import xarray as xr

import matplotlib.pyplot as plt
import matplotlib as mpl
from cartopy import crs as ccrs, feature as cfeature
from cmcrameri import cm
import cmocean as cmo

%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

---

# Load Data

## Phenology stats

```python
def get_phen_info(filename):
    phen_mdl = xr.open_dataset(filename)
    mdl_init = phen_mdl.initiation_ts.astype('<M8[ns]').dt.dayofyear
    mdl_term = phen_mdl.termination_ts.astype('<M8[ns]').dt.dayofyear
    mdl_dura_arr = phen_mdl.duration_ts.values.astype('timedelta64[D]').astype('int')
    mdl_dura_arr =np.ma.masked_where(mdl_dura_arr<0,mdl_dura_arr)
    mdl_phen = xr.Dataset(
            data_vars={
                "init": (['lat','lon'], mdl_init.values),
                "term": (['lat','lon'], mdl_term.values),
                "dura": (['lat','lon'], mdl_dura_arr[:])
            },
            coords = {'lat': (['lat'], phen_mdl.lat.values),
              'lon': (['lon'], phen_mdl.lon.values)
             }
            )
    return mdl_phen
```

```python
phen_mdl = get_phen_info(f"../data/CHL/BIOPERIANT12_PHENOLOGY_clim.nc")
phen_obs = get_phen_info(f"../data/CHL/OCCCIv6_PHENOLOGY_clim.nc")
```

## Plot Phen

```python
# PHEN
month_obs_bounds = [1, 32, 61, 92, 122, 153, 183, 214, 245, 275, 306, 336, 367]
month_mdl_bounds = [1, 31, 59, 90, 120, 151, 181, 212, 243, 273, 304, 334, 365]

cbar_ticks_obs = month_obs_bounds[:12] + np.diff(month_obs_bounds)/2
cbar_ticks_mdl = month_mdl_bounds[:12] + np.diff(month_mdl_bounds)/2
cbar_ticklbls = ['J', 'F', 'M', 'A', 'M', 'J', 'J', 'A', 'S', 'O', 'N', 'D']

cmapit_obs = cmo.tools.lighten(mpl.cm.viridis, 0.8)
normit_obs = mpl.colors.BoundaryNorm(month_obs_bounds, cmapit_obs.N)
cmapit_mdl = cmo.tools.lighten(mpl.cm.viridis, 0.8)
normit_mdl = mpl.colors.BoundaryNorm(month_mdl_bounds, cmapit_mdl.N)

cmapdura = cmo.tools.lighten(mpl.cm.viridis, 0.8)
normdura = mpl.colors.BoundaryNorm(np.arange(0,280,20), cmapdura.N)

fig = plt.figure(figsize=(12, 6))

### ---------MODEL-----JAN---SEP---SCR------------------------------------------- ###
ax = fig.add_subplot(2, 3, 1, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(a) BIOPERIANT12 Initiation',loc='left',fontsize=12)
mesh = ax.pcolormesh(phen_mdl.lon, phen_mdl.lat, phen_mdl.init, 
                     cmap=cmapit_mdl, norm=normit_mdl,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, ticks=cbar_ticks_mdl, orientation='vertical', shrink=0.8, aspect = 25, pad = 0.08)
cb.ax.set_yticklabels(cbar_ticklbls)  
cb.set_label("Month", rotation=-90, labelpad=15)
pu.maps.map_decorator(ax)
pu.maps.add_biomes(ax, dataset="mdl")

ax2 = fig.add_subplot(2, 3, 2, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(b) BIOPERIANT12 Duration',loc='left',fontsize=12)
mesh = ax2.pcolormesh(phen_mdl.lon, phen_mdl.lat, phen_mdl.dura, 
                     cmap=cmapdura, norm=normdura,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, orientation='vertical',drawedges=False)
cb.set_label("Days", rotation=-90, labelpad=15)
pu.maps.map_decorator(ax2)
pu.maps.add_biomes(ax2, dataset="mdl")

ax3 = fig.add_subplot(2, 3, 3, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(c) BIOPERIANT12 Termination', loc='left',fontsize=12)
mesh = ax3.pcolormesh(phen_mdl.lon, phen_mdl.lat, phen_mdl.term, 
                     cmap=cmapit_mdl, norm=normit_mdl,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, ticks=cbar_ticks_mdl, orientation='vertical', shrink=0.8, aspect = 25, pad = 0.08)
cb.ax.set_yticklabels(cbar_ticklbls)  
cb.set_label("Month", rotation=-90, labelpad=15)
pu.maps.map_decorator(ax3)
pu.maps.add_biomes(ax3, dataset="mdl")


### ---------OBS-----JAN---SEP---SCR------------------------------------------- ###
ax4 = fig.add_subplot(2, 3, 4, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(d) OBS Initiation',loc='left',fontsize=12)
mesh = ax4.pcolormesh(phen_obs.lon, phen_obs.lat, phen_obs.init, 
                     cmap=cmapit_obs, norm=normit_obs,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, ticks=cbar_ticks_obs, orientation='vertical', shrink=0.8, aspect = 25, pad = 0.08)
cb.ax.set_yticklabels(cbar_ticklbls)  
cb.set_label("Month", rotation=-90, labelpad=15)
pu.maps.map_decorator(ax4)
pu.maps.add_biomes(ax4, dataset="obs")

ax5 = fig.add_subplot(2, 3, 5, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(e) OBS Duration',loc='left',fontsize=12)
mesh = ax5.pcolormesh(phen_obs.lon, phen_obs.lat, phen_obs.dura, 
                     cmap=cmapdura, norm=normdura,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, orientation='vertical',drawedges=False)
cb.set_label("Days", rotation=-90, labelpad=15)
pu.maps.map_decorator(ax5)
pu.maps.add_biomes(ax5, dataset="obs")


ax6 = fig.add_subplot(2, 3, 6, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(f) OBS Termination',loc='left',fontsize=12)
mesh = ax6.pcolormesh(phen_obs.lon, phen_obs.lat, phen_obs.term, 
                     cmap=cmapit_obs, norm=normit_obs,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, ticks=cbar_ticks_obs, orientation='vertical', shrink=0.8, aspect = 25, pad = 0.08)
cb.ax.set_yticklabels(cbar_ticklbls)  
cb.set_label("Month", rotation=-90, labelpad=15)
pu.maps.map_decorator(ax6)
pu.maps.add_biomes(ax6, dataset="obs")
plt.savefig(f'../figures/S18.png', bbox_inches='tight')
```
