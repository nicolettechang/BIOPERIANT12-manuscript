# Plot temperature and heat content

Source: 
Marcel du Plessis (last update 3 June 2020)  
https://github.com/nicolettechang

### Ocean heat content  

$$H = \int_{0}^x  \rho C_{p} T(z)dz$$
  
$\rho$ = density of seawater  
$C_{p}$ = specific heat capacity
---

# Supp Fig S06 Plot Mean maps

## Plot mean OHC

```python
cmap_ohc = cmo.tools.lighten(plt.cm.hot_r, 0.8) 
bounds = np.arange(440,490,2)
norm_ohc = mpl.colors.BoundaryNorm(bounds, cmap_ohc.N, extend='both')
clev_ohc = [430,440,450,460,470]

######### -------------- PLOT -------------- ###########
fig = plt.figure(figsize=(12, 6))

## Jan

## Jan OHC 
ax = fig.add_subplot(2, 3, 1, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(a) BIOPERIANT12 OHC JAN',loc='left',fontsize=12)
mesh = ax.pcolormesh(ohc_mdl.lon, ohc_mdl.lat, ohc_mdl.isel(time=0), 
                     cmap=cmap_ohc, norm=norm_ohc,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(ohc_label, rotation=-90, labelpad=13)
pu.maps.map_decorator(ax)
pu.maps.add_biomes(ax, dataset="mdl")

ax2 = fig.add_subplot(2, 3, 2, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(b) OBS OHC JAN',loc='left',fontsize=12)
mesh = ax2.pcolormesh(ohc_obs.lon, ohc_obs.lat, ohc_obs[0,:], 
                     cmap=cmap_ohc, norm=norm_ohc,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(ohc_label, rotation=-90, labelpad=13)
pu.maps.map_decorator(ax2)
pu.maps.add_biomes(ax2, dataset="obs")

## SEP OHC 
ax5 = fig.add_subplot(2, 3, 4, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(e) BIOPERIANT12 OHC SEP',loc='left',fontsize=12)
mesh = ax5.pcolormesh(ohc_mdl.lon, ohc_mdl.lat, ohc_mdl.isel(time=8), 
                     cmap=cmap_ohc, norm=norm_ohc,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(ohc_label, rotation=-90, labelpad=13)
pu.maps.map_decorator(ax5)
pu.maps.add_biomes(ax5, dataset="mdl")

ax6 = fig.add_subplot(2, 3, 5, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(f) OBS OHC SEP',loc='left',fontsize=12)
mesh = ax6.pcolormesh(ohc_obs.lon, ohc_obs.lat, ohc_obs[8,:], 
                     cmap=cmap_ohc, norm=norm_ohc,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(ohc_label, rotation=-90, labelpad=13)
pu.maps.map_decorator(ax6)
pu.maps.add_biomes(ax6, dataset="obs")

plt.tight_layout()
plt.savefig(f'../figures/S06a.png', bbox_inches='tight')
```

```python
#SST mean colorbar
cmap_sst= cmo.tools.lighten(cmo.cm.thermal, 0.8) 
boundsT = np.arange(-2,25,1)
norm_sst = mpl.colors.BoundaryNorm(boundsT, cmap_sst.N, extend='both')
clev_sst = [0,10,20]
            
fig = plt.figure(figsize=(12, 6))

## Jan SST
ax3 = fig.add_subplot(2, 3, 1, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(c) BIOPERIANT12 SST JAN',loc='left',fontsize=12)
mesh = ax3.pcolormesh(sst_mdl.lon, sst_mdl.lat, sst_mdl[0,:], 
                     cmap=cmap_sst, norm=norm_sst,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(t_label, rotation=-90, labelpad=10)
pu.maps.map_decorator(ax3)
pu.maps.add_biomes(ax3, dataset="mdl")

ax4 = fig.add_subplot(2, 3, 2, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(d) OBS SST JAN',loc='left',fontsize=12)
mesh = ax4.pcolormesh(sst_obs.lon, sst_obs.lat, sst_obs[0,:], 
                     cmap=cmap_sst, norm=norm_sst,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(t_label, rotation=-90, labelpad=10)
pu.maps.map_decorator(ax4)
pu.maps.add_biomes(ax4, dataset="obs")

## SEP SST
ax7 = fig.add_subplot(2, 3, 4, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(g) BIOPERIANT12 SST SEP',loc='left',fontsize=12)
mesh = ax7.pcolormesh(sst_mdl.lon, sst_mdl.lat, sst_mdl[8,:], 
                     cmap=cmap_sst, norm=norm_sst,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(t_label, rotation=-90, labelpad=10)
pu.maps.map_decorator(ax7)
pu.maps.add_biomes(ax7, dataset="mdl")

ax8 = fig.add_subplot(2, 3, 5, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(h) OBS SST SEP',loc='left',fontsize=12)
mesh = ax8.pcolormesh(sst_obs.lon, sst_obs.lat, sst_obs[8,:], 
                     cmap=cmap_sst, norm=norm_sst,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(t_label, rotation=-90, labelpad=10)
pu.maps.add_biomes(ax8, dataset="obs")
pu.maps.map_decorator(ax8)

plt.tight_layout()
plt.savefig(f'../figures/S06b.png', bbox_inches='tight')
```

```python
from PIL import Image
# Put plots together and save
im0 = Image.open(f'../figures/S06a.png')
im1 = Image.open(f'../figures/S06b.png')
im_final  = Image.new('RGB', (im0.width+im1.width, im0.height), (255, 255, 255))
im_final.paste(im0,(0, 0))
im_final.paste(im1,(im0.width+10, 0))
im_final.show()
im_final.save(f'../figures/S06.png', bbox_inches='tight')
```

# Supp Fig S07 bias and comparison

## Plot model-data bias

```python
# colour bar info
cmap_ohcdiff = cmo.tools.lighten(plt.cm.RdBu_r, 0.8)
bounds = np.arange(-8,9,1)
norm_ohcdiff = mpl.colors.BoundaryNorm(bounds, cmap_ohcdiff.N, extend='both')

cmap_sstdiff = cmo.tools.lighten(plt.cm.RdBu_r, 0.8)
bounds = np.arange(-2,2.2,0.2)
norm_sstdiff = mpl.colors.BoundaryNorm(bounds, cmap_sstdiff.N, extend='both')


######### -------------- PLOT -------------- ###########
fig = plt.figure(figsize=(12, 6))

## Jan
ax = fig.add_subplot(2, 3, 1, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(a) OHC Bias JAN',loc='left',fontsize=12)
mesh = ax.pcolormesh(ohc_obs.lon, ohc_obs.lat, ohc_rgmdl[0,:]-ohc_obs[0,:], 
                     cmap=cmap_ohcdiff, norm=norm_ohcdiff,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(ohc_label, rotation=-90, labelpad=15)
pu.maps.map_decorator(ax)
#pu.add_fronts(ax, 0, dataset="mdl")
pu.maps.add_biomes(ax, dataset="mdl")


ax2 = fig.add_subplot(2, 3, 2, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(b) SST Bias JAN',loc='left',fontsize=12)
mesh = ax2.pcolormesh(sst_obs.lon, sst_obs.lat, sst_rgmdl[0,:]-sst_obs[0,:], 
                     cmap=cmap_sstdiff, norm=norm_sstdiff,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(t_label, rotation=-90, labelpad=15)
pu.maps.map_decorator(ax2)
pu.maps.add_biomes(ax2, dataset="mdl")

## SEP 
ax4 = fig.add_subplot(2, 3, 4, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(c) OHC Bias SEP',loc='left',fontsize=12)
mesh = ax4.pcolormesh(ohc_obs.lon, ohc_obs.lat, ohc_rgmdl[8,:]-ohc_obs[8,:], 
                     cmap=cmap_ohcdiff, norm=norm_ohcdiff,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(ohc_label, rotation=-90, labelpad=15)
pu.maps.map_decorator(ax4)
pu.maps.add_biomes(ax4, dataset="mdl")

ax5 = fig.add_subplot(2, 3, 5, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(d) SST Bias SEP',loc='left',fontsize=12)
mesh = ax5.pcolormesh(sst_obs.lon, sst_obs.lat, sst_rgmdl[8,:]-sst_obs[8,:], 
                     cmap=cmap_sstdiff, norm=norm_sstdiff,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(t_label, rotation=-90, labelpad=15)
pu.maps.map_decorator(ax5)
pu.maps.add_biomes(ax5, dataset="mdl")
plt.savefig(f'../figures/FigS07a.png', bbox_inches='tight')
```

# Plot timeseries

```python
def ax_decoclim(axin):
    axin.set_xlim(0, 11)
    axin.xaxis.set_ticks(np.arange(12))
    axin.xaxis.set_ticklabels(['J', 'A', 'S', 'O', 'N', 'D', 'J', 'F', 'M', 'A', 'M', 'J'])
    axin.grid(zorder=1, lw=1, ls='-', c='0.75', alpha=0.5)
    axin.tick_params(labelsize=10)
    axin.set_rasterized(True)
    return

def ax_decoia(axin):
    axin.set_xlim([pd.Timestamp(1999,12,31), pd.Timestamp(2010,1,1)])
    axin.grid(zorder=1, lw=1, ls='-', c='0.75', alpha=0.5)
    axin.tick_params(labelsize=10)
    axin.set_rasterized(True)
    return

varunit = pu.formatting.get_unit_mplstr(bp12.model_utils.get_varinfo('OHC', 'unit'))
ohc_label = f"OHC {varunit}"

red = pu.formatting.get_rbg_colors('r')
blue = pu.formatting.get_rbg_colors('b')
```

```python
fig = plt.figure(figsize=(6,6))

ax1 = fig.add_subplot(411)
ax1.plot(np.arange(12), np.concatenate((ds_mdl.OHC[6:], ds_mdl.OHC[:6])), red, zorder=20, label='BIOPERIANT12')
ax1.set_ylim(467, 469)
ax1.set_yticks(np.arange(467,470))
ax1.set_ylabel(ohc_label, labelpad=5, color=red, fontsize=10)
ax1.tick_params(axis='y', colors=red)
ax1b = ax1.twinx()
ax1b.plot(np.arange(12), np.concatenate((ds_obs.OHC[6:], ds_obs.OHC[:6])), blue, zorder=20, label='OBS')
ax1b.set_ylim(454, 456)
ax1b.set_yticks(np.arange(454,457))
ax1b.set_ylabel(ohc_label, labelpad=25, color=blue, rotation=-90,fontsize=10)
ax1b.tick_params(axis='y', colors=blue)
pu.timeseries.add_seas(ax1b, (454, 456), is_clim=True, add_legend=False)
ax_decoclim(ax1)
ax1.set_xlabel('Month of year', labelpad=5, fontsize=12)
ax1.set_title('(e) Domain-mean 0-400 m OHC climatology ', loc='left', fontsize=12)
lmdl = mpl.lines.Line2D([], [], color=red, label='BIOPERIANT12')
lobs = mpl.lines.Line2D([], [], color=blue, label='OBS')
leg2 = ax1.legend(handles=[lmdl, lobs], fontsize=10, frameon=True,
                  bbox_to_anchor=(0.1, 1.25), loc="lower left", ncol=2)
ax1.add_artist(leg2)

ax2 = fig.add_subplot(412)
ax2.plot(np.arange(12), np.concatenate((ds_mdl.T0_mean[6:], ds_mdl.T0_mean[:6])), red, zorder=20, label='BIOPERIANT12')
ax2.plot(np.arange(12), np.concatenate((ds_obs.T0_mean[6:], ds_obs.T0_mean[:6])), blue, zorder=20, label='OBS')
ax2.set_ylabel('SST [\u00b0C]', labelpad=5, fontsize=12)
ax2.set_ylim(7, 12)
ax2.set_yticks(np.arange(7,14,2))
ax2.set_title('(f) Domain-mean temperature climatology ', loc='left', fontsize=12)
ax2.set_xlabel('Month of year', labelpad=5, fontsize=12)
ax_decoclim(ax2)
ax2b = ax2.twinx()
ax2b.plot(np.arange(12), np.concatenate((ds_mdl.T200_mean[6:], ds_mdl.T200_mean[:6])), red, linestyle='--', zorder=20, label='BIOPERIANT12')
ax2b.plot(np.arange(12), np.concatenate((ds_obs.T200_mean[6:], ds_obs.T200_mean[:6])), blue, linestyle='--', zorder=20, label='OBS')
ax2b.set_ylim(7., 7.75)
ax2b.set_yticks(np.arange(7., 8, 0.25))
ax2b.set_ylabel('T 200 m [\u00b0C]', labelpad=10, fontsize=10, rotation=-90)
ax2c = ax2.twinx()
ax2c.spines.right.set_position(("axes", 1.17))
ax2c.plot(np.arange(12), np.concatenate((ds_mdl.T400_mean[6:], ds_mdl.T400_mean[:6])), red, linestyle=':', zorder=20, label='BIOPERIANT12')
ax2c.plot(np.arange(12), np.concatenate((ds_obs.T400_mean[6:], ds_obs.T400_mean[:6])), blue, linestyle=':', zorder=20, label='OBS')
ax2c.set_ylim(6, 6.15)
ax2c.set_yticks(np.arange(6., 6.15, 0.05))
pu.timeseries.add_seas(ax2c, (6, 6.15), is_clim=True, add_legend=False)
ax2c.tick_params(direction='in')
ax2c.set_ylabel('T 400 m [\u00b0C]', labelpad=12, fontsize=10, rotation=-90)

ax3 = fig.add_subplot(413)
ax3.plot(ohc_ia_mdl.time, ohc_ia_mdl, 'k', zorder=20, label='BIOPERIANT12')
ax3.set_ylim(467, 468.5)
ytix = np.arange(467,469, 0.25)
ytix_label = [yt if np.mod(i,2)==0 else ' ' for i,yt in enumerate(ytix)]
ax3.set_yticks(ytix)
ax3.set_yticklabels(ytix_label)
ax3.set_ylabel(ohc_label, labelpad=5, fontsize=10)
ax3.set_title('(g) BIOPERIANT12 domain-mean evolution', loc='left', fontsize=12)
ax3.set_xlabel('Year', labelpad=5, fontsize=12)

ax_decoia(ax3)
ax3b = ax3.twinx()
ax3b.tick_params(axis='y', colors=red)
ax3b.plot(t_ia_mdl.time, t_ia_mdl.T0, red, zorder=20)
ax3b.plot(t_ia_mdl.time, t_ia_mdl.T200, red, linestyle='--', zorder=20)
ax3b.plot(t_ia_mdl.time, t_ia_mdl.T400, red, linestyle=':', zorder=20)
ax3b.set_ylim(5, 12)
ax3b.set_yticks(np.arange(5, 12, 1))
ax3b.set_ylabel(t_label, labelpad=10, fontsize=10, color=red, rotation=-90)
pu.timeseries.add_seas(ax3b, (5, 12), is_clim=False, add_legend=False)
lt0   = mpl.lines.Line2D([], [], color=red, label='SST')
lt200 = mpl.lines.Line2D([], [], color=red, linestyle='--', label='T 200 m')
lt400 = mpl.lines.Line2D([], [], color=red, linestyle=':', label='T 400 m')
zleg = ax3b.legend(handles=[lt0, lt200, lt400], fontsize=10, frameon=True,
                  bbox_to_anchor=(0.05,1.25), loc="lower left", ncol=3)
ax3b.add_artist(zleg)
plt.tight_layout()
plt.savefig(f'../figures/FigS07b.png', bbox_inches='tight')
```
