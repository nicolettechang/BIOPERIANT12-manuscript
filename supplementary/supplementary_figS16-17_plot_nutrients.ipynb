# PLOT TIMESERIES 

```python
# biome colours
color17 = pu.formatting.get_biome_colors(17)
color16 = pu.formatting.get_biome_colors(16)
color15 = pu.formatting.get_biome_colors(15) 
```

```python
no3_ylabel = no3_label + '\n solid line'
po4_ylabel = po4_label + '\n dashed line'
si_ylabel = si_label + '\n dotted line'
```

```python
def plot_ts_3ax(no3_in, no3_minmax, po4_in, po4_minmax, si_in, si_minmax, tsclr):
    
    no3_tick, po4_tick, si_tick = no3_minmax[2], po4_minmax[2], si_minmax[2]
    ax.plot(no3_in.time, no3_in.values, label='BIOPERIANT12', c=tsclr, lw=2, alpha=0.7)
    ax.set_xlim([pd.Timestamp(1999,12,31), pd.Timestamp(2010,1,1)])
    ax.set_ylim([no3_minmax[0], no3_minmax[1]])
    ax.set_yticks(np.arange(no3_minmax[0], no3_minmax[1]+no3_tick, no3_tick))
    ax.set_ylabel(no3_ylabel, labelpad=5, fontsize=12)
    _, _, ymin, ymax = ax.axis()
    pu.timeseries.add_seas(ax, (ymin, ymax), False)
    
    ax.grid()
    ax.tick_params(labelsize=12)
    ax.set_rasterized(True)
    ###--- ax2
    ax2 = ax.twinx()
    ax2.plot(po4_in.time, po4_in.values,  c=tsclr, linestyle='--', lw=2)
    ax2.set_ylim([po4_minmax[0], po4_minmax[1]])
    ax2.set_yticks(np.arange(po4_minmax[0], po4_minmax[1]+po4_tick, po4_tick))
    ax2.set_ylabel(po4_ylabel, labelpad=30, fontsize=12, rotation=-90)
    ax2.set_rasterized(True)

    ax3 = ax.twinx()
    ax3.spines.right.set_position(("axes", 1.12))
    ax3.plot(si_in.time, si_in.values, c=tsclr, linestyle=':', lw=3)
    ax3.set_ylim([si_minmax[0], si_minmax[1]])
    ax3.set_yticks(np.arange(si_minmax[0], si_minmax[1]+si_tick, si_tick))
    ax3.tick_params(direction='in')
    ax3.set_ylabel(si_ylabel, labelpad=35, fontsize=12, rotation=-90)
    ax3.set_rasterized(True)
    
def add_var_leg():
    '''add legend for times'''
    no3_l = mpl.lines.Line2D([], [], color=color17, alpha=0.7, label='NO$_3$')
    po4_l = mpl.lines.Line2D([], [], color=color17, linestyle='--', label='PO$_4$')
    si_l  = mpl.lines.Line2D([], [], color=color17, linestyle=':', label='Si')
    legv = ax.legend(handles=[no3_l, po4_l, si_l], fontsize=12, ncol=3,
                   bbox_to_anchor=(0.56, 0.98), loc="lower left", frameon=True)
    ax.add_artist(legv)
```

```python
fig, axs = plt.subplots(4, 6, constrained_layout=True, figsize=(12, 6))
gs = axs[0, 0].get_subplotspec().get_gridspec()
[a.remove() for i in range(6) for a in axs[:, i]]
        
ax = fig.add_subplot(gs[:1, 1:5])

plot_ts_3ax(no3_bts_0_mdl.NO317_mean, (22, 28, 1), po4_bts_0_mdl.PO417_mean, (1.7, 1.95, 0.05), 
            si_bts_0_mdl.Si17_mean, (37, 55, 3), color17)
plt.title(f'(h) SO-ICE BIOPERIANT12 surface',loc='left',fontsize=12)
add_var_leg()

ax = fig.add_subplot(gs[1:2, 1:5])
plot_ts_3ax(no3_bts_0_mdl.NO316_mean, (20, 25, 1), po4_bts_0_mdl.PO416_mean, (1.5, 1.75, 0.05), 
            si_bts_0_mdl.Si16_mean, (21, 31, 2), color16)
plt.title(f'(i) SO-SPSS BIOPERIANT12 surface',loc='left',fontsize=12)


ax = fig.add_subplot(gs[2:3, 1:5])
plot_ts_3ax(no3_bts_0_mdl.NO315_mean, (8, 13, 1), po4_bts_0_mdl.PO415_mean, (0.7, 0.9, 0.05), 
            si_bts_0_mdl.Si15_mean, (4,9,1), color15)
plt.title(f'(j) SO-STSS BIOPERIANT12 surface',loc='left',fontsize=12)
ax.set_xlabel('Year', labelpad=5, fontsize=12)

plt.tight_layout()
plt.savefig(f'../figures/FigS16b.png', bbox_inches='tight')
```
