## Mean biomes from Fay & McKinley 2014, below 30S

### Model mean biome

```python
Model_product = xr.open_dataset(f"../data/BIOMES/BIOPERIANT12onFK2014_biomes_clim_monthly.nc")
meanbiomerg_mdl  = Model_product.biomes
```

### Obs biome

```python
Fay_product = xr.open_dataset("../data/BIOMES/FayMcKinley2014_Biomes.nc")
ds_meanbiomes_obs = Fay_product.MeanBiomes
meanbiomes_obs = ds_meanbiomes_obs.where(ds_meanbiomes_obs.lat<-30, np.nan)
meanbiomes_obs = meanbiomes_obs.transpose()
```

## Find regions where biomes overlap/differ

```python
meanbiome_o = meanbiomes_obs.copy()
b17 = meanbiome_o.where((meanbiomerg_mdl==17) & (meanbiomes_obs==17),0)
b16 = meanbiome_o.where((meanbiomerg_mdl==16) & (meanbiomes_obs==16),0)
b15 = meanbiome_o.where((meanbiomerg_mdl==15) & (meanbiomes_obs==15),0)
b14 = meanbiome_o.where((meanbiomerg_mdl==14) & (meanbiomes_obs==14),0)
b13 = meanbiome_o.where((meanbiomerg_mdl==13) & (meanbiomes_obs==13),0)
b7  = meanbiome_o.where((meanbiomerg_mdl== 7) & (meanbiomes_obs== 7),0)
btot_mean = b17+ b16 +b15 +b14 +b13+ b7
btot_mean = btot_mean.where(btot_mean > 6, np.nan)
```

```python
b17 = meanbiome_o.where((meanbiomerg_mdl!=17) & (meanbiomes_obs==17),0)
b16 = meanbiome_o.where((meanbiomerg_mdl!=16) & (meanbiomes_obs==16),0)
b15 = meanbiome_o.where((meanbiomerg_mdl!=15) & (meanbiomes_obs==15),0)
b14 = meanbiome_o.where((meanbiomerg_mdl!=14) & (meanbiomes_obs==14),0)
b13 = meanbiome_o.where((meanbiomerg_mdl!=13) & (meanbiomes_obs==13),0)
b7  = meanbiome_o.where((meanbiomerg_mdl!= 7) & (meanbiomes_obs== 7),0)
bdiff_mean = b17+ b16 +b15 +b14 +b13+ b7
bdiff_mean = bdiff_mean.where(bdiff_mean > 6, np.nan)
```

### Plot

```python
bounds = np.array([7, 13, 14, 15, 16, 17, 18])
colors_fk = [pu.formatting.get_biome_colors(biomenum) for biomenum in bounds]
cmap_biome = (mpl.colors.ListedColormap(colors_fk))
norm_biome = mpl.colors.BoundaryNorm(bounds, cmap_biome.N)
```

```python
cbar_ticks = [9.5,13.5,14.5,15.5,16.5,17.5]
cbar_ticklbls = ['SP-STPS', 'SA-STPS', 'IND-STPS', 'SO-STSS', 'SO-SPSS', 'SO-ICE']

fig = plt.figure(figsize=(12,6))

######### --- MODEL-BASED MEAN BIOMES (1998-2006) --- ########### 
ax1 = fig.add_subplot(2,3,1, projection = ccrs.Stereographic(central_latitude=-90))
mesh = ax1.pcolormesh(btot_mean.lon, btot_mean.lat, btot_mean, 
                     norm=norm_biome, cmap=cmap_biome, 
                     transform=ccrs.PlateCarree(),zorder=0)
cb = plt.colorbar(mesh, ticks=cbar_ticks, orientation='vertical', shrink=0.8, aspect = 25, pad = 0.08)
cb.ax.set_title('biomes', fontsize=10, fontweight = 'normal', loc='left')
cb.ax.set_yticklabels(cbar_ticklbls)  

plt.title(f'(a) MODEL=OBS (match)', loc='left', fontsize=12)
pu.maps.map_decorator(ax1)
#pu.maps.add_fronts(ax1, -1, dataset="mdl")

######### --- OBSERVATION-BASED CORE BIOMES-- ###########
ax2 = fig.add_subplot(2,3,2, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(b) OBS-MODEL (mismatch)', loc='left', fontsize=12)
mesh2 = ax2.pcolormesh(bdiff_mean.lon, bdiff_mean.lat, bdiff_mean, 
                     norm=norm_biome, cmap=cmap_biome, 
                     transform=ccrs.PlateCarree(),zorder=0)
cb = plt.colorbar(mesh2, ticks=cbar_ticks, orientation='vertical', shrink=0.8, aspect = 25, pad = 0.08)
cb.ax.set_title('biomes', fontsize=10, fontweight = 'normal', loc='left')
cb.ax.set_yticklabels(cbar_ticklbls)  
pu.maps.map_decorator(ax2)
#pu.maps.add_fronts(ax2, -1, dataset="mdl")
plt.savefig(f"../figures/S09.png", dpi=400, bbox_inches='tight')
```

## Plot variables at mismatch

### Get clim

```python
ds = xr.open_dataset("../data/BIOMES/BIOPERIANT12onFK2014_biomeinputs_clim_monthly.nc")
sst_rgmdl = ds.sst
mld_rgmdl = ds.mld
chl_rgmdl = ds.chl
```

### Plot info

```python
varunit = pu.formatting.get_unit_mplstr(bp12.model_utils.get_varinfo('votemper','unit'))
t_label = f" SST {varunit}"
mld_label = f"MLD max [m]"
varunit = pu.formatting.get_unit_mplstr(bp12.model_utils.get_varinfo('chl','unit'))
chl_label = f"Chl-a summer {varunit}"
```

### [ Plot fxn ]

```python
def get_mismatch(biomenum_in):
    if len(biomenum_in) == 1: 
        sstb = sst_rgmdl.where(bdiff_mean==biomenum_in[0])
        mldb = mld_rgmdl.where(bdiff_mean==biomenum_in[0])
        chlb = chl_rgmdl.where(bdiff_mean==biomenum_in[0]) 
    if len(biomenum_in) == 2: 
        sstb = sst_rgmdl.where((bdiff_mean==biomenum_in[0])|(bdiff_mean==biomenum_in[1]))
        mldb = mld_rgmdl.where((bdiff_mean==biomenum_in[0])|(bdiff_mean==biomenum_in[1]))
        chlb = chl_rgmdl.where((bdiff_mean==biomenum_in[0])|(bdiff_mean==biomenum_in[1]))
    if len(biomenum_in) == 3: 
        sstb = sst_rgmdl.where((bdiff_mean==biomenum_in[0])|(bdiff_mean==biomenum_in[1])|(bdiff_mean==biomenum_in[2]))
        mldb = mld_rgmdl.where((bdiff_mean==biomenum_in[0])|(bdiff_mean==biomenum_in[1])|(bdiff_mean==biomenum_in[2]))
        chlb = chl_rgmdl.where((bdiff_mean==biomenum_in[0])|(bdiff_mean==biomenum_in[1])|(bdiff_mean==biomenum_in[2]))
    return sstb, mldb, chlb
            
def plot_vars(biomenum, fignums, sst_lim, chl_lim, mld_lim):
    
    sst_mismatch, mld_mismatch, chl_mismatch = get_mismatch(biomenum)
    
    ax1 = fig.add_subplot(2, 3, 1, projection = ccrs.Stereographic(central_latitude=-90))
    plt.title(f'({fignums[0]}) SST',loc='left',fontsize=12)
    mesh = ax1.pcolormesh(sst_mismatch.lon, sst_mismatch.lat, sst_mismatch,
                         cmap=cm.roma_r, vmin=sst_lim[0], vmax=sst_lim[1],
                         transform=ccrs.PlateCarree(), zorder=0)
    cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
    cb.set_label(t_label, rotation=-90, labelpad=13)
    pu.maps.map_decorator(ax1)
    pu.maps.add_biomes(ax1, "mdl", 0.4)

    ax2 = fig.add_subplot(2, 3, 2, projection = ccrs.Stereographic(central_latitude=-90))
    plt.title(f'({fignums[1]}) Chl-a',loc='left',fontsize=12)
    mesh = ax2.pcolormesh(chl_mismatch.lon, chl_mismatch.lat, chl_mismatch,
                         cmap=cm.roma_r, vmin=chl_lim[0], vmax=chl_lim[1],
                         transform=ccrs.PlateCarree(), zorder=0)
    cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='max')
    cb.set_label(chl_label, rotation=-90, labelpad=13)
    pu.maps.map_decorator(ax2)
    pu.maps.add_biomes(ax2, "mdl", 0.4)
    
    ax3 = fig.add_subplot(2, 3, 3, projection = ccrs.Stereographic(central_latitude=-90))
    plt.title(f'({fignums[2]}) MLD',loc='left',fontsize=12)
    mesh = ax3.pcolormesh(mld_mismatch.lon, mld_mismatch.lat, mld_mismatch,
                         cmap=cm.roma_r, vmin=mld_lim[0], vmax=mld_lim[1],
                         transform=ccrs.PlateCarree(), zorder=0)
    cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='max')
    cb.set_label(mld_label, rotation=-90, labelpad=13)
    pu.maps.map_decorator(ax3)
    pu.maps.add_biomes(ax3, "mdl", 0.4)
    
    return
```

```python
biome = [7,13]
fignums = ("a", "b", "c")
fig = plt.figure(figsize=(12, 6))
plot_vars(biome, fignums,(0,16),(0,0.5),(0,300))
plt.savefig(f"../figures/S10_01.png", dpi=400, bbox_inches='tight')
```
