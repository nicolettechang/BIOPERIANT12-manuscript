## DIC and Alk surface/200m data

```python
ds0_mdl = xr.open_dataset(f"../data/CO2/BIOPERIANT12_DICAlk_0m_clim_monthly.nc")
ds0_obs = xr.open_dataset(f"../data/CO2/OBS_DICAlk_0m_clim_monthly.nc")
```

```python
ds200_mdl = xr.open_dataset(f"../data/CO2/BIOPERIANT12_DICAlk_200m_clim_monthly.nc")
ds200_obs = xr.open_dataset(f"../data/CO2/OBS_DICAlk_200m_clim_monthly.nc")
```

### Plot surface maps

```python
unitstr = pu.formatting.get_unit_mplstr(bp12.model_utils.get_varinfo('DIC','unit'))
dic_cbar = f"DIC {unitstr}"
unitstr = pu.formatting.get_unit_mplstr(bp12.model_utils.get_varinfo('Alkalini','unit'))
alk_cbar = f"Alkalinity {unitstr}"
```

```python
cmapDIC = cmo.tools.crop_by_percent(cm.bamO, 10, which='both', N=None) 
normDIC = mpl.colors.BoundaryNorm(np.arange(2000,2220,20), cmapDIC.N, extend='both')

cmapAlk = cmo.tools.crop_by_percent(cm.bamO, 10, which='both', N=None) 
normAlk = mpl.colors.BoundaryNorm(np.arange(2240,2470,20), cmapAlk.N, extend='both')
```

```python
fig = plt.figure(figsize=(12, 6))

## DIC
ax = fig.add_subplot(2, 3, 1, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(a) BIOPERIANT12 DIC JAN',loc='left',fontsize=12)
mesh = ax.pcolormesh(ds0_mdl.lon, ds0_mdl.lat, ds0_mdl.DIC[0,:], 
                     cmap=cmapDIC, norm=normDIC,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(dic_cbar, rotation=-90, labelpad=13)
pu.maps.map_decorator(ax)
pu.maps.add_biomes(ax, dataset="mdl")

ax2 = fig.add_subplot(2, 3, 2, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(b) BIOPERIANT12 DIC SEP',loc='left',fontsize=12)
mesh = ax2.pcolormesh(ds0_mdl.lon, ds0_mdl.lat, ds0_mdl.DIC[8,:], 
                     cmap=cmapDIC, norm=normDIC,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(dic_cbar, rotation=-90, labelpad=13)
pu.maps.map_decorator(ax2)
pu.maps.add_biomes(ax2, dataset="mdl")


#------------ Alkainity ---------------
ax5 = fig.add_subplot(2, 3, 4, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(e) BIOPERIANT12 TA JAN',loc='left',fontsize=12)
mesh = ax5.pcolormesh(ds0_mdl.lon, ds0_mdl.lat, ds0_mdl.Alkalini[0,:], 
                     cmap=cmapAlk, norm=normAlk,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(dic_cbar, rotation=-90, labelpad=13)
pu.maps.map_decorator(ax5)
pu.maps.add_biomes(ax5, dataset="mdl")

ax6 = fig.add_subplot(2, 3, 5, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(f) BIOPERIANT12 TA SEP',loc='left',fontsize=12)
mesh = ax6.pcolormesh(ds0_mdl.lon, ds0_mdl.lat, ds0_mdl.Alkalini[8,:], 
                     cmap=cmapAlk, norm=normAlk,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(dic_cbar, rotation=-90, labelpad=13)
pu.maps.map_decorator(ax6)
pu.maps.add_biomes(ax6, dataset="mdl")
plt.savefig(f'../figures/S14a.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(12, 6))

## DIC

ax3 = fig.add_subplot(2, 3, 1, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(c) BIOPERIANT12 DIC ANN',loc='left',fontsize=12)
mesh = ax3.pcolormesh(ds0_mdl.lon, ds0_mdl.lat, ds0_mdl.DIC.mean('time'), 
                     cmap=cmapDIC, norm=normDIC,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(dic_cbar, rotation=-90, labelpad=13)
pu.maps.map_decorator(ax3)
pu.maps.add_biomes(ax3, dataset="mdl")

ax4 = fig.add_subplot(2, 3, 2, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(d) OBS DIC ANN',loc='left',fontsize=12)
mesh = ax4.pcolormesh(ds0_obs.lon, ds0_obs.lat, ds0_obs.DIC,  
                     cmap=cmapDIC, norm=normDIC,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(dic_cbar, rotation=-90, labelpad=13)
pu.maps.map_decorator(ax4)
pu.maps.add_biomes(ax4, dataset="obs")


#------------ Alkainity ---------------

ax7 = fig.add_subplot(2, 3, 4, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(g) BIOPERIANT12 TA ANN',loc='left',fontsize=12)
mesh = ax7.pcolormesh(ds0_mdl.lon, ds0_mdl.lat, ds0_mdl.Alkalini.mean('time'), 
                     cmap=cmapAlk, norm=normAlk,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(dic_cbar, rotation=-90, labelpad=13)
pu.maps.map_decorator(ax7)
pu.maps.add_biomes(ax7, dataset="mdl")

ax8 = fig.add_subplot(2, 3, 5, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(h) OBS TA ANN',loc='left',fontsize=12)
mesh = ax8.pcolormesh(ds0_obs.lon, ds0_obs.lat, ds0_obs.Alkalini,  
                     cmap=cmapAlk, norm=normAlk,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(dic_cbar, rotation=-90, labelpad=13)
pu.maps.map_decorator(ax8)
pu.maps.add_biomes(ax8, dataset="obs")

plt.tight_layout()
plt.savefig(f'../figures/S14b.png', bbox_inches='tight')
```

## Timeseries

```python
dic_ts_mdlbiome = xr.open_dataset(f"../data/TIMESERIES/BIOPERIANT12_timeseries_biome_DIC_IA_5d.nc")
alk_ts_mdlbiome = xr.open_dataset(f"../data/TIMESERIES/BIOPERIANT12_timeseries_biome_Alkalini_IA_5d.nc")
```

```python
dic_ts_obsbiome = xr.open_dataset(f"../data/TIMESERIES/OBS_timeseries_DIC_meanbiome_clim.nc")
alk_ts_obsbiome = xr.open_dataset(f"../data/TIMESERIES/OBS_timeseries_TAlk_meanbiome_clim.nc")
```

### PLOT

```python
color15, color16, color17 = [pu.formatting.get_biome_colors(bnum) for bnum in [15,16,17]]
```

```python
varunit = pu.formatting.get_unit_mplstr(bp12.model_utils.get_varinfo('DIC', 'unit'))
dic_ylabel = f"DIC {varunit}"
dic_ylabel_2ax = f"DIC {varunit}\n solid line"

varunit = pu.formatting.get_unit_mplstr(bp12.model_utils.get_varinfo('Alkalini', 'unit'))
alk_ylabel = f"TA {varunit}"
alk_ylabel_2ax = f"TA {varunit}\n dashed line"
```

#### DIC Alk mean surface

```python
fig, axs = plt.subplots(4, 6, constrained_layout=True, figsize=(8, 6))
gs = axs[0, 0].get_subplotspec().get_gridspec()
[a.remove() for i in range(6) for a in axs[:, i]]
        
ax = fig.add_subplot(gs[:1, 1:5])
pu.timeseries.plot_ts_2ax(ax, \
                                dic_ts_mdlbiome.DIC17_mean.sel(deptht=0, method='nearest'), (2120, 2210, 20), \
            alk_ts_mdlbiome.Alkalini17_mean.sel(deptht=0, method='nearest'), (2280, 2330,10), \
            [dic_ylabel, alk_ylabel], [color17, 'k'], [0,1])
plt.title(f'(i) SO-ICE BIOPERIANT12',loc='left',fontsize=12)

pu.timeseries.add_var_leg_2ax(ax, (0.5, 0.95), ['DIC','TA'], [color17, 'k'], [0,1])

ax2 = fig.add_subplot(gs[1:2, 1:5])
pu.timeseries.plot_ts_2ax(ax2, \
                                dic_ts_mdlbiome.DIC16_mean.sel(deptht=0, method='nearest'), (2140, 2160, 5), \
            alk_ts_mdlbiome.Alkalini16_mean.sel(deptht=0, method='nearest'), (2292, 2300, 2), \
            [dic_ylabel, alk_ylabel], [color16, 'k'], [0,1])
plt.title(f'(j) SO-SPSS BIOPERIANT12',loc='left',fontsize=12)

ax3 = fig.add_subplot(gs[2:3, 1:5])
pu.timeseries.plot_ts_2ax(ax3, \
                                dic_ts_mdlbiome.DIC15_mean.sel(deptht=0, method='nearest'), (2070, 2110, 10), \
            alk_ts_mdlbiome.Alkalini15_mean.sel(deptht=0, method='nearest'), (2308, 2316, 2), \
            [dic_ylabel, alk_ylabel], [color15, 'k'], [0,1])                     
plt.title(f'(k) SO-STSS BIOPERIANT12',loc='left',fontsize=12)
ax.set_xlabel('Year', labelpad=5, fontsize=12)
plt.savefig(f'../figures/S14c.png', bbox_inches='tight')
```

## Combine images

```python
from PIL import Image
```

```python
# Full map and timeseries
im1a = Image.open(f'../figures/S14a.png')
im1b = Image.open(f'../figures/S14b.png')
im1c = Image.open(f'../figures/S14c.png')

im_final  = Image.new('RGB', (im1a.width+im1b.width, im1a.height+im1c.height), (255, 255, 255))
im_final.paste(im1a,(0, 0))
im_final.paste(im1b, (im1a.width, 0))
offset = int((im1a.width+im1b.width-im1c.width)/2)
im_final.paste(im1c, (offset, im1a.height))

im_final.show()
im_final.save(f'../figures/S14.png', bbox_inches='tight')
```
