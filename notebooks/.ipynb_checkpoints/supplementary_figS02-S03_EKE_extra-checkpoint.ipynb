# Supplementary: plot zonal and meridional mean EKE

```python
eke_label = f"EKE {pu.formatting_plots.get_unit_mplstr(bp12.model_utils.get_varinfo('eke','unit'))}"
```

## Read averaged EKE 

```python
eke_zon_mdl = xr.open_dataset(f"../data/EKE/BIOPERIANT12_EKE_SOzonal_mean.nc")
eke_mer_mdl = xr.open_dataset(f"../data/EKE/BIOPERIANT12_EKE_SOmerid_mean.nc")
```

```python
eke_zon_obs = xr.open_dataset(f"../data/EKE/AVISO_EKE_SOzonal_mean.nc")
eke_mer_obs = xr.open_dataset(f"../data/EKE/AVISO_EKE_SOmerid_mean.nc")
```

```python
red  = pu.formatting_plots.get_rb_colors('r')
blue = pu.formatting_plots.get_rb_colors('b')
```

```python
fig = plt.figure(figsize=(12,6))

ax = fig.add_subplot(211)
latticks = np.arange(-70, -25, 5)
latlbl = [f"{i}\u00b0 S" for i in np.arange(70, 25, -5) ]
ax.plot(eke_zon_mdl.lat, eke_zon_mdl.eke,color=red, zorder=20,label='BIOPERIANT12')
ax.plot(eke_zon_obs.lat, eke_zon_obs.eke,  color=blue,zorder=20, label='OBS')
ax.grid(zorder=1, lw=1, ls='-', c='0.75', alpha=0.5)
ax.legend(loc=2, ncol=2, fontsize=12)
ax.set_xlim([-70, -30])
ax.set_ylim([0, 400])
plt.title('(a) Zonal-mean EKE',loc ='left',fontsize=12)
ax.set_xticks(latticks)
ax.set_xticklabels(latlbl)
ax.set_xlabel('Latitude', labelpad=5, fontsize=12)
ax.set_ylabel(eke_label, labelpad=10, fontsize=12)
ax.tick_params(labelsize=12)
ax.set_rasterized(True)

## meridional 
lonticks = np.arange(-180,210,30)
lonlbl_w = [f"{i}\u00b0 W" for i in range(180,0,-30)]
lonlbl_e = [f"{i}\u00b0 E" for i in range(30,210,30)]
lonlbl = lonlbl_w + ["0\u00b0 "] + lonlbl_e
ax2 = fig.add_subplot(212)
ax2.plot(eke_mer_mdl.lon, eke_mer_mdl.eke,color=red, zorder=20,label='BIOPERIANT12')
ax2.plot(eke_mer_obs.lon, eke_mer_obs.eke,  color=blue,zorder=20, label='OBS')
ax2.grid(zorder=1, lw=1, ls='-', c='0.75', alpha=0.5)

ax2.legend(loc=2, ncol=2, fontsize=12)
ax2.set_xlim([-180, 180])
ax2.set_ylim([0, 700])
ax2.set_xticks(lonticks)
ax2.set_xticklabels(lonlbl)

pu.map_plots.add_coast(ax2)
ax2.set_xlabel('Longitude', labelpad=5, fontsize=12)
ax2.set_ylabel(eke_label, labelpad=10, fontsize=12)
plt.title('(b) Meridional-mean EKE',loc ='left',fontsize=12)
ax2.tick_params(labelsize=12)
ax2.set_rasterized(True)
plt.tight_layout()

plt.savefig(f'../figures/S02.png', bbox_inches='tight', dpi=300)
```

## Supplementary Figure: Agulhas EKE

### Get annual mean EKE for 2000-2009, BIOPERIANT12 and AVISO

```python
eke_mdl = xr.open_dataset(f"../data/EKE/BIOPERIANT12_EKE_surf_mean.nc").eke
eke_obs = xr.open_dataset(f"../data/EKE/AVISO_EKE_surf_mean.nc").eke
```

### Define region

```python
latmin, latmax, lonmin, lonmax = -44, -32, 5, 45
varin_obs = eke_obs.sel(lon=slice(lonmin-1,lonmax+1), lat=slice(latmin-1,latmax+1))
varin_model = eke_mdl.sel(lon=slice(lonmin-2,lonmax+1), lat=slice(latmin-1,latmax+1))
```

```python
cmapA = cmo.tools.crop_by_percent(cm.vikO, 15, which='both', N=None) 
bounds = np.arange(0,2100,100)
normA = mpl.colors.BoundaryNorm(bounds, cmapA.N, extend='max')

fig = plt.figure(figsize=(20,12))

# MODEL: BIOPERIANT
ax1 = fig.add_subplot(231)
mesh = ax1.pcolormesh(varin_model.lon, varin_model.lat, varin_model, shading='auto',
                     norm=normA, cmap=cmapA,
                     zorder=0)
cb = plt.colorbar(mesh, orientation = 'horizontal', 
                  shrink=0.8, aspect = 25, pad = 0.08, ticks=np.arange(0,2100,400))
cb.set_label(eke_label, labelpad=10)
ax1.set_xlim([lonmin, lonmax])
ax1.set_ylim([latmin, latmax])
ax1.grid(True)
plt.title(f'(a) BIOPERIANT12',loc ='left',fontsize=12)
ax1.set_rasterized(True)

# EKE AVISO OBS
ax2 = fig.add_subplot(232)
mesh = ax2.pcolormesh(varin_obs.lon, varin_obs.lat, varin_obs, shading='auto',
                     norm=normA, cmap=cmapA,
                     zorder=0)
cb = plt.colorbar(mesh, orientation = 'horizontal', 
                  shrink=0.8, aspect = 25, pad = 0.08, ticks=np.arange(0,2100,400))
cb.set_label(eke_label, labelpad=10)

ax2.set_xlim([lonmin, lonmax])
ax2.set_ylim([latmin, latmax])
ax2.grid(True)
plt.title(f'(b) OBS',loc='left',fontsize=12)
ax2.set_rasterized(True)

plt.savefig(f'../figures/S03.png', bbox_inches='tight', dpi=300)
```
