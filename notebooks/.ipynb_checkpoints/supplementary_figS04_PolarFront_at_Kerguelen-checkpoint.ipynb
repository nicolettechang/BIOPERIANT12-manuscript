# Supp: Closer look at Polar Front at the Keguelen Plateau

```python
import sys
import os
src_path = os.path.abspath(os.path.join(os.getcwd(), '../src'))
sys.path.append(src_path)
import bp12_tools as bp12
import bp12_tools.plot_utils as pu

import pandas as pd
import numpy as np
import xarray as xr

import matplotlib.pyplot as plt
import matplotlib as mpl
from cartopy import crs as ccrs, feature as cfeature
from cmcrameri import cm
import cmocean as cmo

%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

## Model IA monthly-mean fronts

```python
ds_mdl_pf  = xr.open_dataset(f"../data/FRONTS/BIOPERIANT12_PF_IA_monthly.nc")
pf_mdl_mean, pf_mdl_std = ds_mdl_pf.lat.mean('time'), ds_mdl_pf.lat.std('time')
```

```python
ds_mdl_saf  = xr.open_dataset(f"../data/FRONTS/BIOPERIANT12_SAF_IA_monthly.nc")
saf_mdl_mean, saf_mdl_std = ds_mdl_saf.lat.mean('time'), ds_mdl_saf.lat.std('time')
```

### OBS WOA Fronts

```python
ds_obs_pf  = xr.open_dataset(f"../data/FRONTS/WOA13_PF_clim_monthly.nc")
pf_obs_mean, pf_obs_std = ds_obs_pf.lat.mean('time'), ds_obs_pf.lat.std('time')
ds_obs_saf = xr.open_dataset(f"../data/FRONTS/WOA13_SAF_clim_monthly.nc")
saf_obs_mean, saf_obs_std = ds_obs_saf.lat.mean('time'), ds_obs_saf.lat.std('time')
```

### OBS Orsi (1995) fronts

```python
ds_orsi_saf = xr.open_mfdataset(f"../data/FRONTS/ORSI_SAF_clim.nc")
ds_orsi_pf  = xr.open_mfdataset(f"../data/FRONTS/ORSI_PF_clim.nc")
```

### Obs Freeman and Lovenduski (2016) Satellite Polar front

```python
satellite_ds = xr.open_dataset(f"../data/FRONTS/Polar_Front_weekly.nc",decode_times=False)
satellite_ds['longitude'] = (((satellite_ds.longitude + 180) % 360) - 180)
satellite_ds = satellite_ds.sortby('longitude')

pf_sat_mean, pf_sat_std = satellite_ds.PFw.mean('time'), satellite_ds.PFw.std('time')
```

### Get model bathy and mask for plot

```python
mask_mdl = bp12.grid_utils.get_bp12grid('tmask').isel(deptht=0)
bathy_mdl = bp12.grid_utils.get_bp12input('Bathymetry', f"../data/GRID/BIOPERIANT12_bathymetry.nc")
```

### Add model EKE

```python
eke_mdl = xr.open_dataset(f"../data/EKE/BIOPERIANT12_EKE_surf_mean.nc").eke
```

```python
eke_label = f"EKE {pu.formatting_plots.get_unit_mplstr(bp12.model_utils.get_varinfo('eke','unit'))}"
```

### Model PF constrain

```python
indx_s = np.where(ds_mdl_pf.lat.sel(lon=slice(70,80)).mean('lon')<-53)[0]
indx_n = np.where(ds_mdl_pf.lat.sel(lon=slice(70,80)).mean('lon')>-53)[0]
```

```python
pf_s_mean, pf_s_std = ds_mdl_pf.lat[indx_s].mean('time'),ds_mdl_pf.lat[indx_s].std('time')
pf_n_mean, pf_n_std = ds_mdl_pf.lat[indx_n].mean('time'),ds_mdl_pf.lat[indx_n].std('time')
```

# Plot  PF pathway over Kergeuelen

### Plot params

```python
# Colormap for mask
cmap = plt.cm.gray
cmap_mask = cmap(np.arange(cmap.N))
cmap_mask[:,-1] = np.linspace(0, 1, cmap.N) * 0.6
cmap_mask = mpl.colors.ListedColormap(cmap_mask)

# Colormap for bathy
cmap = plt.cm.gray
cmap_bathy = cmap(np.arange(cmap.N))
cmap_bathy[:,-1] = np.linspace(0, 1, cmap.N) * 0.2
cmap_bathy = mpl.colors.ListedColormap(cmap_bathy)

## ticks
xticks = np.arange(-180,210,30)
xlbl_w = [f"{i}\u00b0 W" for i in range(180,0,-30)]
xlbl_e = [f"{i}\u00b0 E" for i in range(30,210,30)]
xtl = xlbl_w + ["0\u00b0"] + xlbl_e
xticklabels = []
for i,tl in enumerate(xtl):
    xticklabels.append(tl) if np.mod(i,2)==0 else xticklabels.append("")
        
yticks = np.arange(-65,-35,5)
yticklabels = yticks[:]
yticklabels = [f"{np.abs(tl)}\u00b0 S" for tl in yticks]
```

```python
# cmap for EKE
cmapEKE = cmo.tools.lighten(cmo.cm.balance, 0.8)
bounds = np.concatenate((np.arange(0,100,10), np.arange(100,1100,100)))
normEKE  = mpl.colors.BoundaryNorm(bounds, cmapEKE.N, extend='both')

lonmin, lonmax = 30, 135
latmin, latmax = -65, -40
i1, i2 = bp12.model_utils.find_ind(lonmin, eke_mdl.lon), bp12.model_utils.find_ind(lonmax, eke_mdl.lon),
j1, j2 = bp12.model_utils.find_ind(latmin, eke_mdl.lat), bp12.model_utils.find_ind(latmax, eke_mdl.lat)
lonin, latin =  eke_mdl.lon[i1:i2+1], eke_mdl.lat[j1:j2+1], 
bathyin = bathy_mdl[j1:j2+1, i1:i2+1]
varin_model = eke_mdl[j1:j2+1, i1:i2+1]
```

### Plot

```python
red  = pu.formatting_plots.get_rb_colors('r')
blue = pu.formatting_plots.get_rb_colors('b')
```

```python
fig = plt.figure(figsize=(12, 6))
clabels = [1000,2000]       
gs = mpl.gridspec.GridSpec(5, 6, height_ratios=[1,1,1,1,0.3])
    
ax = fig.add_subplot(gs[:4, :3])
plt.title(f'(a) BIOPERIANT12 SO Indian Basin',loc ='left',fontsize=12)
# Plot EKE
mesh = ax.pcolormesh(lonin, latin, varin_model,
                     norm=normEKE, cmap=cmapEKE,
                     zorder=0)
cb = plt.colorbar(mesh, orientation = 'horizontal', shrink=0.75, aspect = 25, pad = 0.15, extend='max')
cb.set_label(eke_label, labelpad=5, fontsize=10)
# Plot Bathy
CS = plt.contour(lonin,latin,bathyin,clabels, linestyle="--", colors="k", alpha=0.5)
ax.clabel(CS, clabels, inline=True, fontsize=10)

# Plot SAF
ax.plot(saf_mdl_mean.lon, saf_mdl_mean, color=red, linestyle="--", label="SAF", zorder=10)
ax.plot(saf_obs_mean.lon, saf_obs_mean, color=blue, linestyle="--",zorder=10)
ax.plot(ds_orsi_saf.lon, ds_orsi_saf.lat, 'k', linestyle="--",zorder=10)
# PLOT PF

ax.plot(pf_s_mean.lon, pf_s_mean, color=red, lw=2, label="PF", zorder=10)
ax.plot(pf_n_mean.lon, pf_n_mean, color=red, lw=2,  zorder=10)
ax.plot(ds_obs_pf.lon, pf_obs_mean, color=blue, zorder=10)
ax.plot(satellite_ds.longitude, pf_sat_mean, 'g', zorder=10)
ax.plot(ds_orsi_pf.lon,  ds_orsi_pf.lat, 'k', zorder=10)
ax.legend()
ax.set_xlim([lonmin, lonmax])
ax.set_ylim([latmin, latmax])
ax.set_ylabel('Latitude', labelpad=10, fontsize=12)
ax.set_xlabel('Longitude', labelpad=10, fontsize=12)
ax.grid(True)

ax.set_rasterized(True)


#--- RIGHT PANEL ----
ax1 = fig.add_subplot(gs[:2, 3:])
plt.title('(b) BIOPERIANT12 PF Kerguelen north ', loc ='left',fontsize=12)
ax1.plot(pf_n_mean.lon, pf_n_mean, color=red, label="BIOPERIANT12", zorder=10)
for i in indx_n:
    ax1.plot(pf_n_mean.lon, ds_mdl_pf.lat[i],color='grey', alpha=0.2)
ax1.plot(ds_obs_pf.lon, pf_obs_mean, color=blue, label='WOA', zorder=10)
ax1.plot(satellite_ds.longitude, pf_sat_mean, 'g', label='Satellite', zorder=10)
ax1.plot(ds_orsi_pf.lon, ds_orsi_pf.lat, 'k', label='Orsi95')    
#ax1.contourf(bathy_mdl.lon, bathy_mdl.lat, bathy_mdl, [0,3000], cmap=cmap_bathy, zorder=0)
ax1.set_xlim([0,100])
ax1.grid(zorder=1, lw=1, ls='-', c='0.75', alpha=0.5)
#ax1.legend(loc=2, ncol=4, fontsize=12)
ax1.set_xlim([-180, 180])
ax1.set_ylim([-65, -40])
ax1.set_xticks(xticks)
ax1.set_xticklabels(xticklabels)
ax1.set_yticks(yticks)
ax1.set_yticklabels(yticklabels)
ax1.set_ylabel('Latitude', labelpad=10, fontsize=12)
ax1.set_xlabel('Longitude', labelpad=10, fontsize=12)
ax1.tick_params(labelsize=12)
ax1.set_rasterized(True)

ax2 = fig.add_subplot(gs[2:4, 3:])
plt.title('(c) BIOPERIANT12 PF Kerguelen south', loc ='left',fontsize=12)
plt.plot(pf_s_mean.lon, pf_s_mean, color=red,  label="BIOPERIANT12", zorder=10)
#plt.fill_between(pf_s_mean.lon, pf_s_mean-pf_s_std, pf_s_mean+pf_s_std,
#                 alpha=0.2, facecolor='r', zorder=10)
for i in indx_s:
    ax2.plot(pf_n_mean.lon, ds_mdl_pf.lat[i], color='grey', alpha=0.2)
ax2.plot(ds_obs_pf.lon, pf_obs_mean, color=blue, label='WOA', zorder=10)
ax2.plot(satellite_ds.longitude, pf_sat_mean, 'g', label='Satellite', zorder=10)
ax2.plot(ds_orsi_pf.lon,  ds_orsi_pf.lat, 'k', label='Orsi95')
#ax2.contourf(bathy_mdl.lon, bathy_mdl.lat, bathy_mdl, [0,3000], cmap=cmap_bathy, zorder=0)
ax2.grid(zorder=1, lw=1, ls='-', c='0.75', alpha=0.5)
#ax2.legend(ncol=4, fontsize=12, bbox_to_anchor=(-1.3, -0.2), loc="lower left",frameon=True)
ax2.set_xlim([-180, 180])
ax2.set_ylim([-65, -40])
ax2.set_xticks(xticks)
ax2.set_xticklabels(xticklabels)
ax2.set_yticks(yticks)
ax2.set_yticklabels(yticklabels)
ax2.set_ylabel('Latitude', labelpad=10, fontsize=12)
ax2.set_xlabel('Longitude', labelpad=10, fontsize=12)
ax2.tick_params(labelsize=12)
ax2.set_rasterized(True)
plt.tight_layout()

line_bp12 = mpl.lines.Line2D([], [], color=red, lw=2, label='BIOPERIANT12')
line_woa  = mpl.lines.Line2D([], [], color=blue, lw=2, label='WOA13')
line_sat  = mpl.lines.Line2D([], [], color='g', lw=2, label='Satellite')
line_ors  = mpl.lines.Line2D([], [], color='k', lw=2, label='Orsi95')
leg1 = ax2.legend(handles = [line_bp12,line_woa,line_sat,line_ors], ncol=4, fontsize=12, 
                  bbox_to_anchor=(-1.3, -0.3), loc="lower left",frameon=True)
ax2.add_artist(leg1)
plt.savefig(f'../figures/S04.png', bbox_inches='tight', dpi=300)
```
