Source: 
Marcel du Plessis (last update 3 June 2020)  
https://github.com/nicolettechang

### Ocean heat content  

$$H = \int_{0}^x  \rho C_{p} T(z)dz$$
  
$\rho$ = density of seawater  
$C_{p}$ = specific heat capacity

```python
# colour bar info
cmap_ohcdiff = cmo.tools.lighten(plt.cm.RdBu_r, 0.8)
bounds = np.arange(-8,9,1)
norm_ohcdiff = mpl.colors.BoundaryNorm(bounds, cmap_ohcdiff.N, extend='both')

cmap_sstdiff = cmo.tools.lighten(plt.cm.RdBu_r, 0.8)
bounds = np.arange(-2,2.2,0.2)
norm_sstdiff = mpl.colors.BoundaryNorm(bounds, cmap_sstdiff.N, extend='both')


######### -------------- PLOT -------------- ###########
fig = plt.figure(figsize=(12, 6))

## Jan
ax = fig.add_subplot(2, 3, 1, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(a) OHC Bias JAN',loc='left',fontsize=12)
mesh = ax.pcolormesh(ohc_obs.lon, ohc_obs.lat, ohc_rgmdl.sel(month=1)-ohc_obs[0,:], 
                     cmap=cmap_ohcdiff, norm=norm_ohcdiff,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(ohc_label, rotation=-90, labelpad=15)
pu.map_decorator(ax)
pu.add_fronts(ax, 0, dataset="mdl")

ax2 = fig.add_subplot(2, 3, 2, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(b) SST Bias JAN',loc='left',fontsize=12)
mesh = ax2.pcolormesh(sst_obs.lon, sst_obs.lat, sst_rgmdl[0,:]-sst_obs[0,:], 
                     cmap=cmap_sstdiff, norm=norm_sstdiff,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(t_label, rotation=-90, labelpad=15)
pu.map_decorator(ax2)
pu.add_fronts(ax2, 0, dataset="mdl")

## SEP 
ax4 = fig.add_subplot(2, 3, 4, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(c) OHC Bias SEP',loc='left',fontsize=12)
mesh = ax4.pcolormesh(ohc_obs.lon, ohc_obs.lat, ohc_rgmdl.sel(month=9)-ohc_obs[8,:], 
                     cmap=cmap_ohcdiff, norm=norm_ohcdiff,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(ohc_label, rotation=-90, labelpad=15)
pu.map_decorator(ax4)
pu.add_fronts(ax4, 8, dataset="mdl")

ax5 = fig.add_subplot(2, 3, 5, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(d) SST Bias SEP',loc='left',fontsize=12)
mesh = ax5.pcolormesh(sst_obs.lon, sst_obs.lat, sst_rgmdl[8,:]-sst_obs[8,:], 
                     cmap=cmap_sstdiff, norm=norm_sstdiff,
                     transform=ccrs.PlateCarree(), zorder=0)
cb = plt.colorbar(mesh, shrink=0.8, aspect = 25, pad = 0.08, extend='both')
cb.set_label(t_label, rotation=-90, labelpad=15)
pu.map_decorator(ax5)
pu.add_fronts(ax5, 8, dataset="mdl")
plt.savefig(f'{evaldir}/FIGURES/FigS07a.png', bbox_inches='tight', dpi=300)
```

# Plot timeseries

```python
def ax_decoclim(axin):
    axin.set_xlim(0, 11)
    axin.xaxis.set_ticks(np.arange(12))
    axin.xaxis.set_ticklabels(['J', 'A', 'S', 'O', 'N', 'D', 'J', 'F', 'M', 'A', 'M', 'J'])
    axin.grid(zorder=1, lw=1, ls='-', c='0.75', alpha=0.5)
    axin.tick_params(labelsize=10)
    axin.set_rasterized(True)
    return

def ax_decoia(axin):
    axin.set_xlim([pd.Timestamp(1999,12,31), pd.Timestamp(2010,1,1)])
    axin.grid(zorder=1, lw=1, ls='-', c='0.75', alpha=0.5)
    axin.tick_params(labelsize=10)
    axin.set_rasterized(True)
    return

hc_label = f"OHC {pu.get_unit(pu.get_coefs('OHC','unit'))}"

red = pu.get_rb_colors('r')
blue = pu.get_rb_colors('b')
```

```python
fig = plt.figure(figsize=(6,6))

ax1 = fig.add_subplot(411)
ax1.plot(np.arange(12), np.concatenate((ohc_climts_mdl[6:], ohc_climts_mdl[:6])), red, zorder=20, label='BIOPERIANT12')
ax1.set_ylim(467, 469)
ax1.set_yticks(np.arange(467,470))
ax1.set_ylabel(hc_label, labelpad=5, color=red, fontsize=10)
ax1.tick_params(axis='y', colors=red)
ax1b = ax1.twinx()
ax1b.plot(np.arange(12), np.concatenate((ohc_climts_obs[6:], ohc_climts_obs[:6])), blue, zorder=20, label='OBS')
ax1b.set_ylim(454, 456)
ax1b.set_yticks(np.arange(454,457))
ax1b.set_ylabel(hc_label, labelpad=25, color=blue, rotation=-90,fontsize=10)
ax1b.tick_params(axis='y', colors=blue)
pu.add_seas(ax1b, (454, 456), is_clim=True, add_legend=False)
ax_decoclim(ax1)
ax1.set_xlabel('Month of year', labelpad=5, fontsize=12)
ax1.set_title('(e) Domain-mean 0-400 m OHC climatology ', loc='left', fontsize=12)
lmdl = mpl.lines.Line2D([], [], color=red, label='BIOPERIANT12')
lobs = mpl.lines.Line2D([], [], color=blue, label='OBS')
leg2 = ax1.legend(handles=[lmdl, lobs], fontsize=10, frameon=True,
                  bbox_to_anchor=(0.1, 1.25), loc="lower left", ncol=2)
ax1.add_artist(leg2)

ax2 = fig.add_subplot(412)

ax2.plot(np.arange(12), np.concatenate((t0_climts_mdl[6:], t0_climts_mdl[:6])), red, zorder=20, label='BIOPERIANT12')
ax2.plot(np.arange(12), np.concatenate((t0_climts_obs[6:], t0_climts_obs[:6])), blue, zorder=20, label='OBS')
ax2.set_ylabel('SST [\u00b0C]', labelpad=5, fontsize=12)
ax2.set_ylim(7, 12)
ax2.set_yticks(np.arange(7,14,2))
ax2.set_title('(f) Domain-mean temperature climatology ', loc='left', fontsize=12)
ax2.set_xlabel('Month of year', labelpad=5, fontsize=12)
ax_decoclim(ax2)
ax2b = ax2.twinx()
ax2b.plot(np.arange(12), np.concatenate((t200_climts_mdl[6:], t200_climts_mdl[:6])), red, linestyle='--', zorder=20, label='BIOPERIANT12')
ax2b.plot(np.arange(12), np.concatenate((t200_climts_obs[6:], t200_climts_obs[:6])), blue, linestyle='--', zorder=20, label='OBS')
ax2b.set_ylim(7., 7.75)
ax2b.set_yticks(np.arange(7., 8, 0.25))
ax2b.set_ylabel('T 200 m [\u00b0C]', labelpad=10, fontsize=10, rotation=-90)
ax2c = ax2.twinx()
ax2c.spines.right.set_position(("axes", 1.17))
ax2c.plot(np.arange(12), np.concatenate((t400_climts_mdl[6:], t400_climts_mdl[:6])), red, linestyle=':', zorder=20, label='BIOPERIANT12')
ax2c.plot(np.arange(12), np.concatenate((t400_climts_obs[6:], t400_climts_obs[:6])), blue, linestyle=':', zorder=20, label='OBS')
ax2c.set_ylim(6, 6.15)
ax2c.set_yticks(np.arange(6., 6.15, 0.05))
pu.add_seas(ax2c, (6, 6.15), is_clim=True, add_legend=False)
ax2c.tick_params(direction='in')
ax2c.set_ylabel('T 400 m [\u00b0C]', labelpad=12, fontsize=10, rotation=-90)

ax3 = fig.add_subplot(413)
ax3.plot(ohc_iats_mdl.time_counter, ohc_iats_mdl, 'k', zorder=20, label='BIOPERIANT12')
ax3.set_ylim(467, 468.5)
ytix = np.arange(467,469, 0.25)
ytix_label = [yt if np.mod(i,2)==0 else ' ' for i,yt in enumerate(ytix)]
ax3.set_yticks(ytix)
ax3.set_yticklabels(ytix_label)
ax3.set_ylabel(hc_label, labelpad=5, fontsize=10)
ax3.set_title('(g) BIOPERIANT12 domain-mean evolution', loc='left', fontsize=12)
ax3.set_xlabel('Year', labelpad=5, fontsize=12)
ax_decoia(ax3)
ax3b = ax3.twinx()
ax3b.tick_params(axis='y', colors=red)
ax3b.plot(t0_ts_mdl.time, t0_ts_mdl, red, zorder=20)
ax3b.plot(t200_ts_mdl.time, t200_ts_mdl, red, linestyle='--', zorder=20)
ax3b.plot(t400_ts_mdl.time, t400_ts_mdl, red, linestyle=':', zorder=20)
ax3b.set_ylim(5, 12)
ax3b.set_yticks(np.arange(5, 12, 1))
ax3b.set_ylabel(t_label, labelpad=10, fontsize=10, color=red, rotation=-90)
pu.add_seas(ax3b, (5, 12), is_clim=False, add_legend=False)
lt0   = mpl.lines.Line2D([], [], color=red, label='SST')
lt200 = mpl.lines.Line2D([], [], color=red, linestyle='--', label='T 200 m')
lt400 = mpl.lines.Line2D([], [], color=red, linestyle=':', label='T 400 m')
zleg = ax3b.legend(handles=[lt0, lt200, lt400], fontsize=10, frameon=True,
                  bbox_to_anchor=(0.05,1.25), loc="lower left", ncol=3)
ax3b.add_artist(zleg)
plt.tight_layout()
plt.savefig(f'{evaldir}/FIGURES/FigS07b.png', bbox_inches='tight', dpi=300)
```

```python
# biome colours
color17 = pu.get_biome_colors(17)
color16 = pu.get_biome_colors(16)
color15 = pu.get_biome_colors(15) 

def plot_ts(axin, mdlin, obsin, yminmax, tsclr, legpos):
    
    ymin_in, ymax_in, dy_in = yminmax[0], yminmax[1], yminmax[2]
    axin.plot(mdlin.time, mdlin.values, c=tsclr, lw=2)
    axin.plot(obsin.time, obsin.values, c=tsclr, lw=2, linestyle='--')
    axin.set_xlim([pd.Timestamp(1999,12,31), pd.Timestamp(2010,1,1)])
    axin.set_ylim([ymin_in, ymax_in])
    axin.set_yticks(np.arange(ymin_in, ymax_in+dy_in, dy_in))
    axin.set_ylabel(pco2_cbar, labelpad=5, fontsize=12)
    _, _, ymin, ymax = axin.axis()
    pu.add_seas(axin, (ymin, ymax), False)
    
    #add_source(ax)
    #add_var(ax)
    axin.grid()
    axin.tick_params(labelsize=12)
    axin.set_rasterized(True)
    axin.set_xlabel('Year', labelpad=10, fontsize=12)
    line_mdl = mpl.lines.Line2D([], [],  color=tsclr, lw=2, label='BIOPERIANT12')
    line_obs  = mpl.lines.Line2D([], [], color=tsclr, lw=2, linestyle='--', label='OBS')
    leg1 = axin.legend(handles = [line_obs,line_mdl], ncol=2, fontsize=10, 
                  bbox_to_anchor=legpos, loc="lower left",frameon=False)
    axin.add_artist(leg1)
    axin.set_rasterized(True)
    return
   
def plot_pdf(axin, mdlin, obsin, clrin):
    axin.plot(mdlin[0], mdlin[1], c=clrin, lw=2)
    axin.plot(obsin[0], obsin[1], c=clrin, lw=2, linestyle='--') 
    axin.set_xlim(320,400)
    axin.set_xticks(np.arange(320,420,20))
    axin.set_ylim(0, 0.2)
    axin.set_xlabel(pco2_cbar, labelpad=5, fontsize=12)
    axin.set_ylabel("PDF", labelpad=5, fontsize=12)
    axin.grid(lw=1, c='0.75', alpha=0.4)
    axin.set_rasterized(True)  
```

```python
fig = plt.figure(figsize=(6,6))

ax1 = fig.add_subplot(311)
plot_ts(ax1, pco2_ts_mdlbiome.pco217_mean, pco2_ts_obsbiome.pco217_mean, 
        (300, 400, 40), color17, (-0.01, 0.71))
plt.title(f'(g) SO-ICE pCO$_2$ (R={r17[0]:.2f})',loc='left',fontsize=12)


ax2 = fig.add_subplot(312)
plot_ts(ax2, pco2_ts_mdlbiome.pco216_mean, pco2_ts_obsbiome.pco216_mean, 
        (340,380,20), color16, (-0.01, -0.1))
plt.title(f'(i) SO-SPSS pCO$_2$ (R={r16[0]:.2f})',loc='left',fontsize=12)

ax3 = fig.add_subplot(313)
plot_ts(ax3, pco2_ts_mdlbiome.pco215_mean, pco2_ts_obsbiome.pco215_mean, 
        (320,380, 20), color15, (-0.01, -0.1))
plt.title(f'(k) SO-STSS pCO$_2$ (R={r15[0]:.2f})',loc='left',fontsize=12)
plt.tight_layout()
plt.savefig(f'{evaldir}/FIGURES/CO2_Fig1c', bbox_inches='tight', dpi=400)
```
