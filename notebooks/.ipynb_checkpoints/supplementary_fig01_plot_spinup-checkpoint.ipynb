# Plot model spinup 
* domain-avaraged surface EKE  
* Drake Passage Transport

```python
import sys
import os
src_path = os.path.abspath(os.path.join(os.getcwd(), '../src'))
sys.path.append(src_path)
import bp12_tools as bp12
import bp12_tools.plot_utils as pu

import pandas as pd
import numpy as np
import xarray as xr

import matplotlib.pyplot as plt
import matplotlib.dates as mdates

%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

---

## Read data

### Surf EKE evolution 5-day means

```python
ds_eke = xr.open_dataset(f"../data/SPINUP/BIOPERIANT12_timeseries_EKE_surf_IA.nc")
```

```python
eke_label = f"EKE {pu.formatting_plots.get_unit_mplstr(bp12.model_utils.get_varinfo('eke','unit'))}"
```

### Drake Passage transport 5-day means

```python
ds_dpt = xr.open_dataset(f"../data/SPINUP/BIOPERIANT12_timeseries_transport_69W_IA.nc")
```

## Transport

```python
print(f"Annual mean transport 1989-2009 = {ds_dpt.transport.values.mean():5.2f} \u00B1 {ds_dpt.transport.values.std():5.2f} Sv")
```

```python
print(f"Annual mean transport 2000-2009")
transport10 = ds_dpt.transport.sel(time=slice('2000-01-01','2009-12-31'))
print(f"{transport10.values.mean():5.2f} \u00B1 {transport10.values.std():5.2f} Sv")
```

### Calculate monthly means

```python
transport_ts_1m = ds_dpt.resample(time='1M').mean()
transport_ts_1y = ds_dpt.resample(time='1Y').mean()
# Use middle of month/year to plot
transport_ts_1m['time']   =  transport_ts_1m.time - pd.Timedelta('15 days')
transport_ts_1y['time']   =  transport_ts_1y.time - pd.Timedelta('182 days')
```

# PLOT

```python
red = pu.formatting_plots.get_rb_colors('r')
```

```python
fig = plt.figure(figsize=(12,4))

# --------------- EKE SPIN UP --------------- #
ax = fig.add_subplot(211)
ax.plot(ds_eke.time, ds_eke.eke.values, zorder=20, c=red)
ax.grid(zorder=1, lw=1, ls='-', c='0.75', alpha=0.5)
ax.set_xlim([pd.Timestamp(1988, 12, 31), pd.Timestamp(2010, 1, 1)])
ax.set_ylim([0, 200])
ax.set_yticks([0,50,100,150,200])
ax.xaxis.set_major_locator(mdates.YearLocator(1) )
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.set_xlabel('Year', labelpad=5, fontsize=12)
ax.set_ylabel(eke_label, labelpad=10, fontsize=12)
plt.title('(a) BIOPERIANT12  EKE surface domain-averaged',loc ='left',fontsize=12)
ax.tick_params(labelsize=12)
ax.set_rasterized(True)

# --------------- Transport SPIN UP --------------- #
ax2 = fig.add_subplot(212)
ax2.plot(transport_ts_1m.time, transport_ts_1m.transport.values, color=red, label='Monthly mean')
ax2.plot(transport_ts_1y.time, transport_ts_1y.transport.values, 'k--',   label='Annual mean')
ax2.grid(zorder=1, lw=1, ls='-', c='0.75', alpha=0.5)
ax2.legend(loc=1, ncol=2, fontsize=12)
ax2.set_xlim([pd.Timestamp(1988, 12, 31), pd.Timestamp(2010, 1, 1)])
ax2.set_ylim([130, 175])
ax2.set_yticks([130,140,150,160,170])
ax2.xaxis.set_major_locator(mdates.YearLocator(1) )
ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax2.set_xlabel('Year', labelpad=5, fontsize=12)
ax2.set_ylabel('Transport [Sv]', labelpad=10, fontsize=12)
plt.title('(b) BIOPERIANT12 Drake Passage transport',loc ='left',fontsize=12)
ax2.tick_params(labelsize=12)
ax2.set_rasterized(True)

plt.tight_layout()
plt.savefig(f'../figures/S01.png', bbox_inches='tight', dpi=300)
```
