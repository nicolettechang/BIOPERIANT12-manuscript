# Plot biome maps

---

Source: Alice Lebehot
"Here, we define biogeochemical-physical boundaries in the Southern Ocean using Fay & McKinley 2014. As in their publication, I defined the time-varying (yearly) biomes and the mean biome over the period of study. However we had to make some adjustments to the coastal areas in the model outputs, which show an ice fraction < 0.5 (little ice) around Antarctica's coast - which is unlikely. Those points were removed."
---

```python
import sys
import os
src_path = os.path.abspath(os.path.join(os.getcwd(), '../src'))
sys.path.append(src_path)
import bp12_tools as bp12
import bp12_tools.plot_utils as pu

import xarray as xr
import numpy as np

import matplotlib.colors as colors
import matplotlib.pyplot as plt
from cartopy import crs as ccrs, feature as cfeature
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

---

# Mean biomes from Fay & McKinley 2014, below 30S

```python
Fay_product = xr.open_dataset(f'../data/BIOMES/FayMcKinley2014_Biomes.nc')
ds_meanbiomes_obs = Fay_product.MeanBiomes
meanbiomes_obs = ds_meanbiomes_obs.where(ds_meanbiomes_obs.lat<-30, np.nan)
meanbiomes_obs = meanbiomes_obs.transpose()
```

### Make discrete color map

```python
bounds = np.array([7, 13, 14, 15, 16, 17, 18])
colors_fk = [pu.formatting.get_biome_colors(biomenum) for biomenum in bounds]
cmap_biome = (colors.ListedColormap(colors_fk))
norm_biome = colors.BoundaryNorm(bounds, cmap_biome.N)
```

## Mean biomes in Southern Hemisphere

```python
SO_biomes = ds_meanbiomes_obs.where(meanbiomes_obs.lat<-10, np.nan).transpose()
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(1,2,1, projection = ccrs.Stereographic(central_latitude=-90))
ax.gridlines(alpha=1, linewidth=0.35, linestyle='--', ylocs=np.arange(-80,0,10))
ax.set_global()
ax.set_extent([-180, 180, -90, -30], ccrs.PlateCarree())   
plt.title('(a) OBS', loc='left', fontsize=12)
 
mesh = ax.pcolormesh(SO_biomes.lon, SO_biomes.lat, SO_biomes, 
                     norm=norm_biome, cmap=cmap_biome, vmax=18,
                     transform=ccrs.PlateCarree(),zorder=0)
ax.coastlines('110m', lw=1.5)
cb = plt.colorbar(mesh) 
cb.ax.set_title('biomes',fontsize=10,fontweight = 'normal')
plt.tight_layout()
```

Biomes 7, 13, 14, 15, 16 and 17 exist in Fay & McKinley 2014 in the Southern Ocean. So I need to define them in BIOPERIANT12.

Biome 7 = South Pacific Subtropical Permanently Stratified = SP-STPS   
Biome 13 = South Atlantic Subtropical Permanently Stratified = SA-STPS  
Biome 14 = Indian Ocean Subtropical Permanently Stratified = IND STPS  
Biome 15 = Southern Ocean Subtropical Seasonally Stratified = SO-STSS  
Biome 16 =Southern Ocean Subpolar Seasonally Stratified = SO-SPSS  
Biome 17 =Southern Ocean Ice = SO-ICE

### Model mean biome

```python
Model_product = xr.open_dataset(f"../data/BIOMES/BIOPERIANT12_biomes_clim_monthly.nc")
meanbiomes_model = Model_product.biomes
```

### Get area of grid using lat and lon

```python
e1t = bp12.grid_utils.get_bp12grid('e1t')
e2t = bp12.grid_utils.get_bp12grid('e2t')
area_model = e1t * e2t 
```

```python
area_obs = xr.DataArray(data=bp12.grid_utils.make_obs_areagrid(meanbiomes_obs.lon, meanbiomes_obs.lat),
                        dims=["lat", "lon"], 
                        coords=[meanbiomes_obs.lat.values, meanbiomes_obs.lon.values])
```

# Plot biome map

```python
cbar_ticks = [9.5,13.5,14.5,15.5,16.5,17.5]
cbar_ticklbls = ['SP-STPS', 'SA-STPS', 'IND-STPS', 'SO-STSS', 'SO-SPSS', 'SO-ICE']

fig = plt.figure(figsize=(16,9))

######### --- MODEL-BASED MEAN BIOMES (1998-2006) --- ########### 
ax1 = fig.add_subplot(1,3,1, projection = ccrs.Stereographic(central_latitude=-90))
mesh = ax1.pcolormesh(meanbiomes_model.lon, meanbiomes_model.lat, meanbiomes_model, 
                     norm=norm_biome, cmap=cmap_biome, 
                     transform=ccrs.PlateCarree(),zorder=0)
cb = plt.colorbar(mesh, ticks=cbar_ticks, orientation='vertical', shrink=0.4, aspect = 25, pad = 0.08)
cb.ax.set_title('biomes', fontsize=10, fontweight = 'normal', loc='left')
cb.ax.set_yticklabels(cbar_ticklbls)  
plt.title(f'(a) BIOPERIANT12', loc='left', fontsize=12)
pu.maps.map_decorator(ax1)
pu.maps.add_fronts(ax1, -1, dataset="mdl")

######### --- OBSERVATION-BASED CORE BIOMES-- ###########
ax2 = fig.add_subplot(1,3,2, projection = ccrs.Stereographic(central_latitude=-90))
plt.title('(b) OBS', loc='left', fontsize=12)
mesh2 = ax2.pcolormesh(meanbiomes_obs.lon, meanbiomes_obs.lat, meanbiomes_obs, 
                     norm=norm_biome, cmap=cmap_biome, 
                     transform=ccrs.PlateCarree(),zorder=0)
cb = plt.colorbar(mesh2, ticks=cbar_ticks, orientation='vertical', shrink=0.4, aspect = 25, pad = 0.08)
cb.ax.set_title('biomes', fontsize=10, fontweight = 'normal', loc='left')
cb.ax.set_yticklabels(cbar_ticklbls)  
pu.maps.map_decorator(ax2)
pu.maps.add_fronts(ax2, -1, dataset="obs")

plt.tight_layout()
plt.savefig(f'../figures/Fig6a_biomes.png', bbox_inches='tight')
```

# Compare mean biomes

## Biome size

```python
area7_mean_model  = area_model.where(meanbiomes_model==7).sum(('lat','lon'))
area13_mean_model = area_model.where(meanbiomes_model==13).sum(('lat','lon'))
area14_mean_model = area_model.where(meanbiomes_model==14).sum(('lat','lon'))
area15_mean_model = area_model.where(meanbiomes_model==15).sum(('lat','lon'))
area16_mean_model = area_model.where(meanbiomes_model==16).sum(('lat','lon'))
area17_mean_model = area_model.where(meanbiomes_model==17).sum(('lat','lon'))
```

```python
area7_mean_obs = area_obs.where(meanbiomes_obs==7).sum(('lat','lon'))
area13_mean_obs = area_obs.where(meanbiomes_obs==13).sum(('lat','lon'))
area14_mean_obs = area_obs.where(meanbiomes_obs==14).sum(('lat','lon'))
area15_mean_obs = area_obs.where(meanbiomes_obs==15).sum(('lat','lon'))
area16_mean_obs = area_obs.where(meanbiomes_obs==16).sum(('lat','lon'))
area17_mean_obs = area_obs.where(meanbiomes_obs==17).sum(('lat','lon'))
```

```python
fig, ax = plt.subplots(figsize=(4,4))
coef = 1e-12
#-------------------
barwidth = 0.2 #for i in range(12)]
mdl7 = ax.bar(0.25, coef*area7_mean_model.values,
               width=barwidth, color=colors_fk[0], 
            hatch='\\\\', edgecolor='k')
mdl13 = ax.bar(0.25 + (1*0.7), coef*area13_mean_model.values,
               width=barwidth, color=colors_fk[1], 
            hatch='\\\\', edgecolor='k')
mdl14 = ax.bar(0.25 + (2*0.7), coef*area14_mean_model.values,
               width=barwidth, color=colors_fk[2],
            hatch='\\\\', edgecolor='k')
mdl15 = ax.bar(0.25 + (3*0.7), coef*area15_mean_model.values,
               width=barwidth, color=colors_fk[3], 
            hatch='\\\\', edgecolor='k')
mdl16 = ax.bar(0.25 + (4*0.7), coef*area16_mean_model.values, 
               width=barwidth, color=colors_fk[4], label='BIOPERIANT12',
            hatch='\\\\', edgecolor='k')
mdl17 = ax.bar(0.25 + (5*0.7), coef*area17_mean_model.values, 
               width=barwidth, color=colors_fk[5], 
            hatch='\\\\', edgecolor='k')

obs7 = ax.bar(0.5, coef*area7_mean_obs.values,
               width=barwidth, color=colors_fk[0], edgecolor='k')
obs13 = ax.bar(0.5 + (1*0.7), coef*area13_mean_obs.values, 
               width=barwidth, color=colors_fk[1], edgecolor='k')
obs14 = ax.bar(0.5 + (2*0.7), coef*area14_mean_obs.values,  
               width=barwidth, color=colors_fk[2], edgecolor='k')
obs15 = ax.bar(0.5 + (3*0.7), coef*area15_mean_obs.values,
               width=barwidth, color=colors_fk[3], edgecolor='k')
obs16 = ax.bar(0.5 + (4*0.7), coef*area16_mean_obs.values, label='OBS',
               width=barwidth, color=colors_fk[4], edgecolor='k')
obs17 = ax.bar(0.5 + (5*0.7), coef*area17_mean_obs.values, 
               width=barwidth, color=colors_fk[5], edgecolor='k')


ax.set_ylabel("Area [10$^6$ km$^2$]", labelpad=10, fontsize=12)
ax.legend()
ax.xaxis.set_ticks([])
ax.xaxis.set_ticks(np.arange(0.3, 4, 0.7))
plt.tick_params(bottom=False)
ax.xaxis.set_ticklabels(['SP-STPS', 'SA-STPS', 'IND-STPS', 'SO-STSS', 
                       'SO-SPSS', 'SO-ICE'], fontsize=10, rotation=45)  
ax.set_title('(c) Total area of mean biomes', loc='left', fontsize=12)
ax.text(-0.05,  coef*area7_mean_model.values+1.5, 'BP12', fontsize=9)
ax.text(0.3, coef*area7_mean_obs.values+1.5, 'OBS', fontsize=9)
plt.tight_layout()

plt.savefig(f'../figures/Fig6b_biomes.png', bbox_inches='tight')
```
